## 1. FUNDAMENTOS TEÓRICOS

### 1.1 Definición de Seguridad de la Información
La seguridad de la información es el conjunto de medidas preventivas y reactivas de las organizaciones y sistemas tecnológicos que permiten resguardar y proteger la información buscando mantener la confidencialidad, disponibilidad e integridad de la misma.

#### Conceptos Fundamentales:
- **Información**: Todo aquel conjunto de datos organizados en poder de una entidad que posean valor para la misma.
- **Seguridad**: Estado en el cual los riesgos y las amenazas son conocidos y gestionados hasta un nivel aceptable.
- **Sistema de Gestión**: Marco de políticas, procedimientos, directrices y recursos asociados para lograr los objetivos de la organización.

### 1.2 Principios Básicos (CIA Triad)

#### Confidencialidad
- **Definición**: Propiedad que garantiza que la información no se pone a disposición ni se revela a individuos, entidades o procesos no autorizados.
- **Ejemplos**:
  - Datos personales de clientes
  - Información financiera
  - Secretos comerciales
- **Medidas típicas**:
  - Cifrado de datos
  - Control de accesos
  - Acuerdos de confidencialidad

#### Integridad
- **Definición**: Propiedad de exactitud y completitud de los activos.
- **Ejemplos**:
  - Registros contables
  - Códigos fuente
  - Documentos legales
- **Medidas típicas**:
  - Firmas digitales
  - Control de versiones
  - Checksums

#### Disponibilidad
- **Definición**: Propiedad de estar accesible y utilizable cuando lo requiera una entidad autorizada.
- **Ejemplos**:
  - Servicios web críticos
  - Sistemas de emergencia
  - Bases de datos operativas
- **Medidas típicas**:
  - Sistemas redundantes
  - Copias de seguridad
  - Planes de continuidad

### 1.3 Principios Extendidos

#### No Repudio
- **Definición**: Capacidad de demostrar la ocurrencia de un evento reclamado y sus entidades originadoras.
- **Implementación**: Firmas digitales, sellos de tiempo, logs auditables.

#### Autenticidad
- **Definición**: Propiedad que garantiza que una entidad es quien dice ser.
- **Mecanismos**: Certificados digitales, tokens de seguridad, biometría.

# Marcos Regulatorios de Protección de Datos y Privacidad

##  RGPD (Reglamento General de Protección de Datos) en Colombia

El Reglamento General de Protección de Datos (RGPD) es una normativa de la Unión Europea diseñada para proteger los datos personales de los ciudadanos europeos. Sin embargo, aunque Colombia no está directamente obligada a cumplir con el RGPD, existen puntos de conexión importantes entre esta normativa y las leyes de protección de datos en Colombia, como la Ley 1581 de 2012 y el Decreto 1377 de 2013, que regulan la protección de datos personales en el país.

### Relación entre el RGPD y la normativa colombiana

1. **Principios compartidos**:  
   Tanto el RGPD como la Ley 1581 de 2012 en Colombia comparten principios clave, como:
   - **Consentimiento informado**: Los datos personales deben ser tratados con el consentimiento previo, expreso e informado del titular.
   - **Finalidad específica**: Los datos solo pueden ser usados para los fines explícitos por los cuales fueron recolectados.
   - **Seguridad**: Se deben implementar medidas para garantizar la protección de los datos personales frente a accesos no autorizados o pérdidas.

2. **Transferencia internacional de datos**:  
   El RGPD regula estrictamente la transferencia de datos fuera de la UE. Colombia, aunque no es miembro de la UE, es reconocida por tener un nivel adecuado de protección de datos personales, lo que facilita el intercambio de información bajo ciertas condiciones.

3. **Responsabilidad y derechos de los titulares**:  
   Ambas normativas garantizan derechos similares a los titulares, como:
   - Derecho a acceder a los datos personales.
   - Derecho a la rectificación, actualización o eliminación de los datos.
   - Derecho a oponerse al tratamiento de los datos.

4. **Obligaciones de las empresas**:  
   En el RGPD se menciona la figura del "responsable del tratamiento" y el "encargado del tratamiento", roles similares a los descritos en la Ley 1581. Las empresas en Colombia deben garantizar que sus prácticas de manejo de datos cumplan con estándares adecuados, especialmente si tienen relaciones comerciales con países de la UE.

### Diferencias clave
- **Alcance territorial del RGPD**: Este reglamento se aplica a cualquier empresa en el mundo que trate datos personales de ciudadanos de la UE. Por otro lado, la Ley 1581 se aplica exclusivamente al territorio colombiano, salvo cuando los datos sean transferidos o tratados internacionalmente.
- **Multas y sanciones**: Las multas bajo el RGPD pueden ser significativamente más altas (hasta 20 millones de euros o el 4% de la facturación anual). En Colombia, las sanciones de la Superintendencia de Industria y Comercio (SIC) son menores, pero aún pueden ser sustanciales.

### Conexión práctica: Empresas en Colombia que tratan datos de la UE
Si una empresa colombiana trata datos de ciudadanos de la UE (por ejemplo, en comercio electrónico o servicios digitales), debe cumplir con los requisitos del RGPD además de la normativa local. Esto incluye:
   - Nombrar un representante dentro de la UE.
   - Garantizar el derecho al olvido, la portabilidad de datos y otras disposiciones del RGPD.


La **Ley 1581 de 2012**

Es la normativa principal en Colombia que regula la protección de datos personales. Su objetivo es garantizar el derecho fundamental a la **privacidad**, permitiendo que las personas tengan control sobre el tratamiento de su información personal.

### Aspectos principales de la Ley 1581 de 2012

#### **1. Ámbito de Aplicación**
La ley aplica a:
- Personas naturales (titulares de datos) cuyos datos personales se recolectan o procesan.
- Empresas, organizaciones o personas (responsables y encargados del tratamiento) que manejen datos personales en Colombia, independientemente de si el tratamiento se realiza dentro o fuera del país.

Excepciones:
- Bases de datos con fines exclusivamente personales o domésticos.
- Bases de datos reguladas por leyes específicas, como las relacionadas con la seguridad nacional, inteligencia y contrainformación.

---

#### **2. Principios Rectores**
La ley establece principios fundamentales para el manejo de datos personales:
- **Legalidad**: Todo tratamiento debe respetar las leyes vigentes.
- **Finalidad**: Los datos deben recolectarse con un propósito específico y legítimo.
- **Libertad**: Los datos solo pueden tratarse con el consentimiento previo, expreso e informado del titular.
- **Transparencia**: El titular debe tener acceso a su información en cualquier momento.
- **Seguridad**: Se deben implementar medidas para proteger los datos contra accesos no autorizados o usos indebidos.
- **Confidencialidad**: Los datos no pueden divulgarse sin autorización previa o sin una base legal.

---

#### **3. Derechos de los Titulares**
Los ciudadanos tienen derechos específicos sobre sus datos personales:
1. **Acceso**: Solicitar información sobre cómo se están usando sus datos.
2. **Rectificación**: Actualizar o corregir información incorrecta.
3. **Cancelación**: Solicitar la eliminación de datos cuando ya no sean necesarios o se haya retirado el consentimiento.
4. **Oposición**: Negarse al tratamiento de sus datos en casos específicos.
5. **Prueba del consentimiento**: Pedir evidencia de que el responsable obtuvo el consentimiento adecuado.

---

#### **4. Responsabilidades del Responsable y Encargado del Tratamiento**
- **Responsable del Tratamiento**: Define los fines y medios para el manejo de los datos.
- **Encargado del Tratamiento**: Realiza el tratamiento de datos en nombre del responsable.

Ambos deben:
- Garantizar que se sigan los principios de la ley.
- Implementar medidas de seguridad para proteger los datos.
- Atender las solicitudes de los titulares de datos.

---

#### **5. Registro Nacional de Bases de Datos (RNBD)**
- Las empresas o entidades que tratan datos personales deben registrar sus bases de datos en el RNBD administrado por la **Superintendencia de Industria y Comercio (SIC)**.
- El registro permite mayor control y transparencia en el manejo de los datos personales.

---

#### **6. Sanciones**
La **Superintendencia de Industria y Comercio (SIC)** es la autoridad encargada de supervisar el cumplimiento de la Ley 1581. Puede imponer sanciones que incluyen:
- Multas de hasta 2.000 salarios mínimos legales mensuales vigentes (SMLMV).
- Suspensión temporal de actividades relacionadas con el tratamiento de datos.
- Cierre definitivo de operaciones si se encuentra incumplimiento grave.

---

### Decreto Reglamentario 1377 de 2013
Este decreto complementa la Ley 1581 y detalla aspectos operativos:
- Proporciona reglas sobre el manejo de datos sensibles.
- Establece plazos y procedimientos para solicitar el consentimiento a los titulares.
- Regula el manejo de datos de menores de edad, exigiendo que su tratamiento respete sus derechos fundamentales.

---

### Relación con Normativas Internacionales
Aunque la Ley 1581 se enfoca en el contexto colombiano, tiene similitudes con el **Reglamento General de Protección de Datos (RGPD)** de la Unión Europea. Esto permite que empresas colombianas puedan adaptar sus políticas para el tratamiento de datos de ciudadanos extranjeros, especialmente en mercados internacionales.

## 3. Comparativa RGPD vs CCPA

### 3.1 Tabla Comparativa
| Aspecto | RGPD | CCPA |
|---------|------|------|
| Ámbito Territorial | Global (si aplica) | California |
| Base Legal | 6 bases legales | No requiere |
| Consentimiento | Explícito | Opt-out |
| Multas | Hasta 4% facturación | $7,500 por violación |
| Notificación Breach | 72 horas | No específico |
| DPO | Obligatorio | No requerido |

### 3.2 Principales Diferencias
1. **Alcance**
   - RGPD: Más amplio y detallado
   - CCPA: Específico para California

2. **Definiciones**
   - RGPD: Datos personales
   - CCPA: Información personal

3. **Obligaciones**
   - RGPD: Más prescriptivo
   - CCPA: Más flexible
