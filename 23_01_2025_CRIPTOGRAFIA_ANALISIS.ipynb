# Criptografía
La criptografía es la práctica de proteger la información mediante técnicas de cifrado y descifrado. El objetivo principal es garantizar la confidencialidad, integridad y autenticidad de los datos.

Tipos de criptografía
1. Criptografía simétrica: Utiliza la misma llave para cifrar y descifrar los datos. Ejemplos: AES, DES.
2. Criptografía asimétrica: Utiliza una llave pública para cifrar y una llave privada para descifrar. Ejemplos: RSA, Elliptic Curve.
3. Criptografía de hash: Utiliza una función de hash para generar un resumen digital único de los datos. Ejemplos: SHA-256, MD5.

Algoritmos criptográficos
1. AES (Advanced Encryption Standard): Algoritmo de cifrado simétrico ampliamente utilizado.
2. RSA (Rivest-Shamir-Adleman): Algoritmo de cifrado asimétrico utilizado para autenticación y cifrado.
3. Elliptic Curve: Algoritmo de cifrado asimétrico utilizado para autenticación y cifrado.

# Autenticación
La autenticación es el proceso de verificar la identidad de un usuario, dispositivo o sistema.

Tipos de autenticación
1. Autenticación de contraseña: Utiliza una contraseña para verificar la identidad.
2. Autenticación biométrica: Utiliza características biométricas (huellas dactilares, reconocimiento facial) para verificar la identidad.
3. Autenticación de dos factores: Utiliza una combinación de dos factores (contraseña y token, por ejemplo) para verificar la identidad.

Protocolos de autenticación
1. Kerberos: Protocolo de autenticación utilizado para redes de área local (LAN).
2. RADIUS (Remote Authentication Dial-In User Service): Protocolo de autenticación utilizado para redes de área amplia (WAN).
3. OAuth: Protocolo de autenticación utilizado para aplicaciones web y móviles.

Introducción a la criptografía y autenticación

La criptografía es la práctica de proteger la información mediante técnicas de cifrado y descifrado. La autenticación es el proceso de verificar la identidad de un usuario, dispositivo o sistema.

Conceptos básicos de criptografía

- Cifrado: El proceso de convertir la información en un formato que no pueda ser entendido sin la clave adecuada.
- Descifrado: El proceso de convertir la información cifrada en un formato legible.
- Clave: La información secreta utilizada para cifrar y descifrar la información.
- Hash: Un resumen digital único de la información.

Uso de la biblioteca hashlib para generar hashes de archivos y cadenas de texto

La biblioteca hashlib es una biblioteca de Python que proporciona una forma fácil de generar hashes de archivos y cadenas de texto.

Ejercicio de script básico




```python
# Generar un hash de un archivo utilizando hashlib.


import hashlib

# Seleccionar el archivo que se desea generar el hash
archivo = "mi_archivo.txt"

# Seleccionar el algoritmo de hash que se desea utilizar
algoritmo = "sha256"

# Generar el hash del archivo
with open(archivo, "rb") as f:
    hash = hashlib.sha256(f.read()).hexdigest()

print(f"El hash del archivo {archivo} es: {hash}")
```

```python
# Generar una firma digital utilizando hmac.


import hmac
import hashlib

# Seleccionar la clave secreta que se desea utilizar
clave = "mi_clave_secreta"

# Seleccionar el mensaje que se desea firmar
mensaje = "mi_mensaje"

# Generar la firma digital del mensaje
firma = hmac.new(clave.encode(), mensaje.encode(), hashlib.sha256).hexdigest()

print(f"La firma digital del mensaje {mensaje} es: {firma}")
```

```python
# Verificar la integridad de un archivo utilizando hashlib y hmac.


import hashlib
import hmac

# Seleccionar el archivo que se desea verificar
archivo = "mi_archivo.txt"

# Seleccionar la clave secreta que se desea utilizar
clave = "mi_clave_secreta"

# Generar el hash del archivo
with open(archivo, "rb") as f:
    hash = hashlib.sha256(f.read()).hexdigest()

# Generar la firma digital del hash
firma = hmac.new(clave.encode(), hash.encode(), hashlib.sha256).hexdigest()

print(f"El hash del archivo {archivo} es: {hash}")
print(f"La firma digital del hash es: {firma}")
```

Nivel básico: Uso de la biblioteca cryptography para cifrar y descifrar archivos y cadenas de texto

La biblioteca cryptography es una biblioteca de Python que proporciona una forma segura y fácil de cifrar y descifrar archivos y cadenas de texto.

Instalación de la biblioteca cryptography

```python
Cifrar un archivo utilizando cryptography y AES.

from cryptography.hazmat.primitives import padding
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend

# Seleccionar el archivo que se desea cifrar
archivo = "mi_archivo.txt"

# Seleccionar la clave de cifrado
clave = "mi_clave_de_cifrado"

# Cifrar el archivo
with open(archivo, "rb") as f:
    datos = f.read()

cipher = Cipher(algorithms.AES(clave.encode()), modes.CBC(b'\0' * 16), backend=default_backend())
cifrador = cipher.encryptor()
datos_cifrados = cifrador.update(datos) + cifrador.finalize()

# Guardar los datos cifrados en un archivo
with open("mi_archivo_cifrado.txt", "wb") as f:
    f.write(datos_cifrados)
```

```python
Descifrar un archivo utilizando cryptography y AES.

from cryptography.hazmat.primitives import padding
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend

# Seleccionar el archivo que se desea descifrar
archivo = "mi_archivo_cifrado.txt"

# Seleccionar la clave de descifrado
clave = "mi_clave_de_cifrado"

# Descifrar el archivo
with open(archivo, "rb") as f:
    datos_cifrados = f.read()

cipher = Cipher(algorithms.AES(clave.encode()), modes.CBC(b'\0' * 16), backend=default_backend())
descifrador = cipher.decryptor()
datos = descifrador.update(datos_cifrados) + descifrador.finalize()

# Guardar los datos descifrados en un archivo
with open("mi_archivo_descifrado.txt", "wb") as f:
    f.write(datos)
```
