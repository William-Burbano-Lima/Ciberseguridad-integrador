# **Diccionario de 100 Ciberdelitos comunes**

1. **Phishing**:
Es una técnica de ingeniería social que engaña a las víctimas para que revelen información personal, como contraseñas o números de tarjetas de crédito, generalmente mediante correos electrónicos o sitios web falsos que aparentan ser de confianza.

2. **Malware**:
Es un término genérico que abarca todo tipo de software malicioso diseñado para dañar, interrumpir o acceder de forma no autorizada a sistemas informáticos.

3. **Ransomware**:
Tipo de malware que encripta los archivos de la víctima y exige un rescate para restaurar el acceso a ellos.

4. **DDoS (Distributed Denial of Service)**:
Ataque que sobrecarga un servidor o red con una gran cantidad de peticiones, lo que impide el acceso legítimo a los usuarios.

5. **Spyware**:
Software que recopila información sobre una persona u organización sin su conocimiento y la envía a otra entidad sin su consentimiento.

6. **Hacking**: Actividad que implica acceder a sistemas informáticos de manera no autorizada, a menudo con el objetivo de robar datos o causar daños.

7. **Robo de Identidad**: Consiste en obtener y utilizar ilegalmente información personal de otra persona para cometer fraude o delitos.

8. **Cyberstalking**: Uso de internet u otros medios digitales para acosar o amenazar a una persona de manera persistente y maliciosa.

9. **Carding**: Uso de tarjetas de crédito robadas para realizar transacciones no autorizadas, a menudo en pequeñas cantidades para evitar la detección.

10. **Exploit**: Secuencia de comandos o software que aprovecha una vulnerabilidad de seguridad para ejecutar código malicioso en un sistema.

11. **Keylogger**: Herramienta que registra las teclas presionadas en un teclado, utilizada para capturar información confidencial como contraseñas.

12. **Ingeniería Social**: Manipulación psicológica de las personas para que realicen acciones o divulguen información confidencial, sin necesidad de habilidades técnicas avanzadas.

13. **Botnets**: Red de computadoras infectadas por malware, controladas de forma remota por un atacante para llevar a cabo actividades maliciosas, como ataques DDoS.

14. **Crypto-jacking**: Uso no autorizado de los recursos de computadoras ajenas para minar criptomonedas, generalmente mediante malware.

15. **Violación de Datos (Data Breach)**: Incidente en el que información sensible o confidencial es accedida, robada o divulgada sin autorización.

16. **Ataque de Día Cero (Zero-Day Attack)**: Explotación de una vulnerabilidad de software desconocida para el proveedor, dejándolos sin tiempo para desarrollar un parche.

17. **Ciberacoso (Cyberbullying)**: Uso de tecnología digital para acosar o intimidar a una persona de manera repetida y hostil.

18. **Ataque de Intermediario (MitM)**: Tipo de ataque donde un atacante intercepta y posiblemente altera la comunicación entre dos partes sin su conocimiento.

19. **Cracking**: Proceso de romper la seguridad de un software para eliminar protecciones como la gestión de derechos digitales (DRM) o la activación de software.

20. **Vishing**: Variante del phishing que utiliza llamadas telefónicas para engañar a las víctimas y obtener información confidencial.

21. **Smishing**: Phishing a través de mensajes SMS, donde los atacantes envían mensajes para engañar a las víctimas y que compartan información personal.

22. **Pharming**: Manipulación de las direcciones DNS para redirigir el tráfico de un sitio web legítimo a uno falso controlado por atacantes.

23. **Skimming Digital**: Robo de información de tarjetas de crédito mediante dispositivos colocados en lectores de tarjetas o sitios web comprometidos.

24. **Shoulder Surfing**: Observación de personas mientras ingresan información confidencial, como contraseñas o números de tarjetas de crédito.

25. **Salami Slicing**: Estrategia que implica robar pequeñas cantidades de dinero o recursos que pasan desapercibidas, pero que acumulativamente resultan en grandes sumas.

26. **Bluejacking**: Envio de mensajes no solicitados a dispositivos habilitados con Bluetooth dentro del rango del atacante.

27. **Bluebugging**: Acceso no autorizado a un dispositivo Bluetooth para escuchar llamadas o leer mensajes de texto.

28. **Juacking (Video Jacking)**: Uso de estaciones de carga USB manipuladas para robar datos de dispositivos conectados.

29. **Ataques de Diccionario**: Método de crackeo de contraseñas que prueba una lista de posibles combinaciones, generalmente palabras comunes o previamente utilizadas.

30. **Ataques de Fuerza Bruta**: Proceso de prueba y error para adivinar contraseñas o claves cifradas probando todas las combinaciones posibles.

31. **Ataques de Repetición (Replay Attacks)**: Intercepción y retransmisión de datos de comunicación con el objetivo de engañar al receptor.

32. **Ataques de Relleno de Credenciales (Credential Stuffing)**: Uso de combinaciones de usuario y contraseña robadas de otros sitios para acceder a cuentas en diferentes servicios.

33. **Desfiguración de Sitios Web (Website Defacement)**: Alteración no autorizada del contenido de un sitio web, generalmente para mostrar mensajes políticos o despectivos.

34. **Secuestro de Sesiones (Session Hijacking)**: Intercepción de una sesión activa de un usuario para tomar el control de la sesión sin su conocimiento.

35. **Inyección SQL**: Ataque que consiste en insertar código SQL malicioso en una entrada de usuario para manipular la base de datos subyacente.

36. **Inyección de Comandos (Command Injection)**: Explotación de vulnerabilidades en la validación de entradas para ejecutar comandos arbitrarios en el servidor.

37. **Inyección de Código (Code Injection)**: Inserción de código malicioso en una aplicación vulnerable para alterar su funcionamiento o comprometer el sistema.

38. **Ataques XSS (Cross-Site Scripting)**: Inyección de scripts maliciosos en sitios web confiables para ejecutar código en el navegador de los usuarios.

39. **CSRF (Cross-Site Request Forgery)**: Ataque que fuerza a un usuario autenticado a ejecutar acciones no deseadas en una aplicación web en la que está autenticado.

40. **Clickjacking**: Engaño a los usuarios para que hagan clic en algo diferente de lo que creen, generalmente con el objetivo de robar información o realizar acciones no autorizadas.

41. **Path Traversal**: Vulnerabilidad que permite a los atacantes acceder a archivos y directorios fuera del directorio previsto por la aplicación.

42. **Ataques de Overflow de Buffer (Buffer Overflow)**: Consiste en enviar más datos a un buffer de los que puede manejar, lo que puede resultar en la ejecución de código malicioso.

43. **Escalamiento de Privilegios**: Obtención de derechos de acceso más altos en un sistema, más allá de los que debería tener un usuario ordinario.

44. **Ataques Man-in-the-Browser**: Variante del ataque MitM donde el malware se inyecta en el navegador web para interceptar y manipular las comunicaciones entre el usuario y el sitio web.

45. **Secuestro de DNS (DNS Hijacking)**: Manipulación de los servidores DNS para redirigir el tráfico de un dominio legítimo a uno malicioso.

46. **Eavesdropping**: Escucha no autorizada de comunicaciones privadas, como llamadas telefónicas o mensajes electrónicos.

47. **Packet Sniffing**: Captura y análisis de paquetes de datos que se transmiten por una red, utilizada para recopilar información como contraseñas o datos confidenciales.

48. **Ataques de Desbordamiento de Pila (Stack Overflow)**: Tipo de buffer overflow donde el desbordamiento ocurre en la pila de memoria, lo que puede permitir la ejecución de código malicioso.

49. **Ataques de Desbordamiento de Heap (Heap Overflow)**: Similar al desbordamiento de pila, pero ocurre en la sección de la memoria heap, utilizada para la asignación dinámica.

50. **Robo de Cookies**: Obtención no autorizada de cookies de sesión para acceder a cuentas o información personal del usuario.

51. **Malvertising**: Uso de anuncios en línea maliciosos para distribuir malware.

52. **Ataques de Watering Hole**: Compromiso de sitios web frecuentados por una organización o grupo objetivo para infectar a sus usuarios.

53. **Ataques de Rootkit**: Software diseñado para ocultarse en el sistema y permitir el acceso no autorizado a los atacantes.

54. **Ataques de Bootkit**: Variante de rootkit que infecta el sector de arranque de un disco para controlar el sistema desde el inicio.

55. **Ataques de Click Fraud**: Generación de clics falsos en anuncios en línea para inflar los costos o ingresos de publicidad.

56. **Ataques de Typosquatting**: Registro de dominios con nombres similares a sitios populares para engañar a los usuarios y robar información.

57. **Ataques de Cybersquatting**: Registro de nombres de dominio con la intención de venderlos a las empresas o personas legítimas.

58. **Ataques de Domain Hijacking**: Robo de control de un nombre de dominio mediante técnicas como phishing o hacking.

59. **Ataques de Firmware**: Compromiso de firmware de dispositivos para permitir acceso no autorizado o control.

60. **Ataques de Hardware**: Manipulación física de dispositivos para insertar backdoors o malware.

61. **Ataques de Side Channel**: Recolección de información a través de la observación de los canales secundarios como el consumo de energía o el tiempo de ejecución.

62. **Ataques de Cryptanalysis**: Intentos de descifrar la encriptación sin la clave adecuada mediante métodos matemáticos.

63. **Ataques de Ransomware Dirigido**: Infección de sistemas específicos con ransomware para extorsionar a organizaciones o personas.

64. **Ataques de Spear Phishing**: Phishing altamente personalizado dirigido a individuos o grupos específicos.

65. **Ataques de Whale Phishing**: Tipo de spear phishing dirigido a altos ejecutivos o personas influyentes dentro de una organización.

66. **Ataques de Clone Phishing**: Creación de una copia de un correo electrónico legítimo previamente enviado para engañar a los destinatarios.

67. **Ataques de Phishing Suplantando Identidades Famosas**: Phishing que se hace pasar por personas o entidades conocidas para ganar la confianza del objetivo.

68. **Ataques de Drive-by Download**: Descarga automática de software malicioso al visitar un sitio web comprometido.

69. **Ataques de Cryptolocker**: Variante de ransomware que cifra los archivos de la víctima y exige un rescate para su liberación.

70. **Ataques de Ransomware de Pantalla Completa**: Ransomware que bloquea la pantalla del dispositivo y exige un rescate para desbloquearla.

71. **Ataques de Scareware**: Software malicioso que intenta asustar a los usuarios para que descarguen software no deseado o paguen por una solución falsa.

72. **Ataques de Rogueware**: Software que se hace pasar por una aplicación legítima para engañar al usuario e instalar malware.

73. **Ataques de Secuestro de Cámaras**: Acceso no autorizado a la cámara de un dispositivo para espiar o grabar al usuario.

74. **Ataques de Secuestro de Micrófonos**: Acceso no autorizado al micrófono de un dispositivo para espiar conversaciones.

75. **Ataques de Secuestro de Dispositivos IoT**: Control no autorizado de dispositivos IoT para diversos fines maliciosos.

76. **Ataques de Cryptographic Sprawl**: Descontrol en la gestión de claves y certificados criptográficos, aumentando el riesgo de brechas de seguridad.

77. **Ataques de Password Spraying**: Intento de adivinar contraseñas mediante el uso de las contraseñas más comunes en múltiples cuentas.

78. **Ataques de Tabnabbing**: Engaño en el que se modifica una pestaña del navegador inactiva para parecerse a una página de inicio de sesión legítima.

79. **Ataques de Dumpster Diving**: Búsqueda de información sensible en la basura o desechos de una organización.

80. **Ataques de Trashing**: Similar al dumpster diving, pero se enfoca en la búsqueda de dispositivos electrónicos desechados.

81. **Ataques de WiFi Jamming**: Interferencia intencional con señales WiFi para interrumpir las comunicaciones.

82. **Ataques de WiFi Spoofing**: Configuración de redes WiFi falsas para engañar a los usuarios y capturar su información.

83. **Ataques de Rogue Access Point**: Creación de puntos de acceso WiFi no autorizados para interceptar el tráfico de red.

84. **Ataques de Evil Twin**: Punto de acceso WiFi malicioso que imita uno legítimo para engañar a los usuarios.

85. **Ataques de Warchalking**: Marcado de lugares públicos donde se encuentran redes WiFi abiertas o vulnerables.

86. **Ataques de Warchalking con Bluetooth**: Similar al warchalking, pero se enfoca en redes y dispositivos Bluetooth.

87. **Ataques de Fingerprinting**: Recolección de información específica del dispositivo o usuario para rastrear o identificar sin su consentimiento.

88. **Ataques de Waterline**: Compromiso de datos almacenados en almacenamiento persistente, como discos duros.

89. **Ataques de Watermarking**: Inserción de marcas en documentos digitales para rastrear su uso o distribución.

90. **Ataques de Troyano de Acceso Remoto (RAT)**: Software que permite al atacante controlar un sistema de manera remota.

91. **Ataques de Troyano Bancario**: Troyanos diseñados específicamente para robar información financiera de usuarios.

92. **Ataques de Troyano para Exfiltración de Datos**: Troyanos que extraen información confidencial de un sistema.

93. **Ataques de Ransomware Móvil**: Variante de ransomware que afecta dispositivos móviles.

94. **Ataques de Ransomware IoT**: Ransomware que infecta dispositivos IoT, bloqueando su funcionalidad hasta que se pague un rescate.

95. **Ataques de Manipulación de Precios**: Alteración ilegal de precios en plataformas de comercio electrónico.

96. **Ataques de Manipulación de Mercado**: Actividades ilegales para influir en el precio o el volumen de acciones en el mercado.

97. **Ataques de Spam**: Envío masivo de correos electrónicos no deseados, generalmente con fines publicitarios o fraudulentos.

98. **Ataques de Spoofing de Correo Electrónico**: Técnica para falsificar la dirección de remitente en correos electrónicos.

99. **Ataques de Spoofing de Identidad**: Suplantación de identidad mediante el uso de datos personales falsos.

100. **Ataques de Suplantación de Voz (Voice Spoofing)**: Uso de grabaciones o síntesis de voz para hacerse pasar por otra persona durante llamadas telefónicas.
