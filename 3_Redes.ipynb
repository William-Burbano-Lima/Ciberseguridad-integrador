### **Sección 1: Conceptos Básicos de Redes**

- **Red de Datos**: Un sistema de interconexión de dispositivos que permite la transmisión de datos entre ellos. Las redes pueden ser físicas o inalámbricas.
- **Tipos de Redes**:
  - **LAN (Local Area Network)**: Red que conecta dispositivos en un área geográfica limitada, como una oficina o un edificio.
  - **WAN (Wide Area Network)**: Red que abarca grandes áreas geográficas, como ciudades o países, interconectando múltiples LANs.
  - **MAN (Metropolitan Area Network)**: Red que cubre una ciudad o área metropolitana.
  - **PAN (Personal Area Network)**: Red de corto alcance, como la que conecta dispositivos personales.
- **Componentes de una Red**:
  - **Router**: Dispositivo que dirige paquetes de datos entre redes diferentes.
  - **Switch**: Dispositivo que conecta múltiples dispositivos en una LAN y gestiona el flujo de datos dentro de la red.
  - **Hub**: Dispositivo que conecta varios dispositivos en una red, pero a diferencia de un switch, envía datos a todos los puertos.
  - **NIC (Network Interface Card)**: Tarjeta de hardware que permite a un dispositivo conectarse a una red.

---

### **Sección 2: Protocolo de Internet y Cómo Viajan los Paquetes**

- **Modelo OSI (Open Systems Interconnection)**: Modelo conceptual con 7 capas que define cómo los datos deben viajar a través de una red. Las capas son: Física, Enlace de Datos, Red, Transporte, Sesión, Presentación, y Aplicación.
- **Modelo TCP/IP**: Modelo más práctico que se utiliza en internet, con 4 capas: Acceso a la Red, Internet, Transporte, y Aplicación.
- **Dirección IP (Internet Protocol)**:
  - **IPv4**: Sistema de direccionamiento de 32 bits con direcciones formadas por cuatro octetos.
  - **IPv6**: Sistema de 128 bits diseñado para superar las limitaciones de IPv4.
- **Subnetting**: Proceso de dividir una red IP en subredes más pequeñas para optimizar el uso de direcciones IP.
- **Enrutamiento**: Proceso por el cual un router decide la ruta óptima para enviar un paquete de datos hacia su destino.

---

### **Sección 3: Protocolos de Internet**

- **TCP (Transmission Control Protocol)**: Protocolo de capa de transporte que asegura la entrega de datos completos y en orden.
- **UDP (User Datagram Protocol)**: Protocolo de transporte más simple que no garantiza la entrega ni el orden, pero es más rápido que TCP.
- **HTTP/HTTPS (Hypertext Transfer Protocol/Secure)**: Protocolos utilizados para la transferencia de páginas web, con HTTPS ofreciendo una capa de seguridad adicional.
- **FTP (File Transfer Protocol)**: Protocolo para la transferencia de archivos entre sistemas en una red.
- **SMTP (Simple Mail Transfer Protocol)**: Protocolo para el envío de correos electrónicos.
- **DNS (Domain Name System)**: Sistema que traduce nombres de dominio legibles por humanos (como www.ejemplo.com) en direcciones IP numéricas.
- **ARP (Address Resolution Protocol)**: Protocolo utilizado para mapear direcciones IP a direcciones MAC en una red local.

---

### **Sección 4: Arquitectura de Internet y Ciberseguridad Básica**

- **Arquitectura Cliente-Servidor**: Modelo donde los clientes solicitan servicios y recursos a un servidor central.
- **Arquitectura Peer-to-Peer (P2P)**: Modelo donde todos los dispositivos en la red tienen capacidades equivalentes y pueden actuar como clientes o servidores.
- **Amenazas Comunes en Redes**:
  - **DDoS (Distributed Denial of Service)**: Ataque que intenta hacer que un servicio no esté disponible mediante la sobrecarga del servidor con tráfico.
  - **Spoofing**: Técnica donde un atacante engaña a una red haciéndose pasar por otro dispositivo.
  - **Sniffing**: Proceso de interceptar y registrar el tráfico de la red.
- **Medidas Básicas de Seguridad en Redes**:
  - **Firewall**: Dispositivo o software que controla el acceso a la red basado en reglas de seguridad.
  - **VPN (Virtual Private Network)**: Conexión segura a través de una red pública que permite a los usuarios enviar y recibir datos de manera privada.
  - **Cifrado**: Proceso de codificar datos para evitar que terceros los lean.

**Simulación con Python:**
- Crear un script que simule el bloqueo de IPs usando un firewall en Python:

```python
import socket

# Crear un servidor básico
def create_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 8080))
    server_socket.listen(5)
    print("Esperando conexiones...")
    
    while True:
        client_socket, client_address = server_socket.accept()
        print(f"Conexión establecida con {client_address}")
        
        # Simulando un firewall que bloquea una IP específica
        if client_address[0] == "192.168.1.100":  # Bloquear esta IP
            print(f"Conexión rechazada desde {client_address[0]}")
            client_socket.close()
        else:
            client_socket.send(b'Conexión permitida')
            client_socket.close()

# Ejecutar servidor
create_server()
```

### **Simulación de Red Básica con Python**

Simularemos cómo dos dispositivos se comunican a través de un servidor y un cliente.

```python
import socket

# Crear un servidor básico
def crear_servidor():
    servidor_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    servidor_socket.bind(('localhost', 8080))
    servidor_socket.listen(1)
    print("Esperando conexión...")
    cliente_socket, cliente_direccion = servidor_socket.accept()
    print(f"Conexión establecida con {cliente_direccion}")
    cliente_socket.send(b'Hola, cliente')
    cliente_socket.close()

# Crear un cliente que se conecta al servidor
def crear_cliente():
    cliente_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    cliente_socket.connect(('localhost', 8080))
    data = cliente_socket.recv(1024)
    print(f"Recibido: {data.decode()}")
    cliente_socket.close()

# Ejecutamos el servidor y cliente (en diferentes celdas o entornos)
crear_servidor()  # Ejecutar esto en un entorno de servidor
crear_cliente()   # Ejecutar esto en el cliente
```

## **1. Configuración del Firewall en Linux (iptables)**

### **Paso 1: Comprobar el estado de iptables**
Para asegurarte de que `iptables` esté instalado y funcionando, puedes ejecutar el siguiente comando:
```bash
sudo systemctl status netfilter-persistent
```

Si `iptables` no está instalado, instálalo con:
```bash
sudo apt-get install iptables
```

### **Paso 2: Bloquear el acceso a un puerto específico**
Supongamos que queremos bloquear el acceso al puerto 8080 (por ejemplo, para un servidor web que está corriendo en ese puerto).

Ejecuta el siguiente comando para bloquear el puerto 8080:
```bash
sudo iptables -A INPUT -p tcp --dport 8080 -j DROP
```

Este comando hace lo siguiente:
- `-A INPUT`: Añadir una regla a la cadena de entrada (INPUT).
- `-p tcp`: Especificar el protocolo TCP.
- `--dport 8080`: Especifica el puerto 8080.
- `-j DROP`: Indica que el tráfico hacia ese puerto debe ser bloqueado.

### **Paso 3: Permitir el acceso a ciertos servicios (por ejemplo, HTTP)**
Ahora, supongamos que quieres permitir el acceso solo al servicio HTTP (puerto 80). Usa este comando:
```bash
sudo iptables -A INPUT -p tcp --dport 80 -j ACCEPT
```

Esto permitirá el tráfico solo hacia el puerto 80 (HTTP), mientras que todos los demás puertos estarán bloqueados, incluida la redirección del tráfico hacia el puerto 8080.

### **Paso 4: Verificar las reglas de iptables**
Para verificar las reglas actuales de `iptables`, puedes usar:
```bash
sudo iptables -L
```

Esto te mostrará las reglas activas en tu sistema.

### **Paso 5: Guardar las reglas (opcional)**
Si quieres que estas reglas se mantengan después de reiniciar el sistema, puedes guardarlas con el siguiente comando:
```bash
sudo iptables-save > /etc/iptables/rules.v4
```

## **2. Configuración del Firewall en Windows (Windows Firewall)**

### **Paso 1: Acceder a Windows Firewall**
1. Abre el **Panel de Control** y selecciona **Firewall de Windows**.
2. Haz clic en **Configuración avanzada** en el panel izquierdo. Esto abrirá la ventana de **Firewall de Windows con seguridad avanzada**.

### **Paso 2: Bloquear un puerto específico**
1. En la ventana de **Firewall de Windows con seguridad avanzada**, selecciona **Reglas de entrada**.
2. Haz clic en **Nueva regla** en el panel derecho.
3. En la ventana de **Asistente para nueva regla de entrada**, selecciona **Puerto** y haz clic en **Siguiente**.
4. Selecciona **TCP** y luego **Puertos locales específicos**. Introduce el puerto que deseas bloquear, por ejemplo, `8080`, y haz clic en **Siguiente**.
5. En la siguiente pantalla, selecciona **Bloquear la conexión** y haz clic en **Siguiente**.
6. Asigna un nombre a la regla, como "Bloquear puerto 8080", y haz clic en **Finalizar**.

### **Paso 3: Permitir acceso a ciertos servicios**
Para permitir el acceso a ciertos servicios como HTTP (puerto 80), sigue los mismos pasos anteriores pero selecciona **Permitir la conexión** en lugar de **Bloquear la conexión**.

1. En el paso 4, elige el puerto 80 para HTTP.
2. En el paso 5, selecciona **Permitir la conexión** y haz clic en **Siguiente**.

### **Paso 4: Verificar las reglas del firewall**
Puedes ver las reglas activas en **Reglas de entrada** y **Reglas de salida** en la ventana de **Firewall de Windows con seguridad avanzada**.

### **Introducción a Puertos: Qué Son y Cómo Identificar Vulnerabilidades**

#### **¿Qué son los Puertos?**
En redes de computadoras, un **puerto** es un punto de acceso a un dispositivo o red que se utiliza para recibir o enviar datos. Los puertos son identificados por números, y cada número de puerto está asociado a un servicio o aplicación específica.

Los puertos pueden clasificarse en tres categorías:
1. **Puertos bien conocidos (0-1023)**: Son utilizados por servicios y aplicaciones que se usan de forma común. Ejemplos incluyen:
   - **Puerto 80**: HTTP (servicio web).
   - **Puerto 443**: HTTPS (HTTP seguro).
   - **Puerto 25**: SMTP (envío de correos electrónicos).
   - **Puerto 22**: SSH (acceso remoto seguro).
   - **Puerto 21**: FTP (transferencia de archivos).

2. **Puertos registrados (1024-49151)**: Estos puertos pueden ser utilizados por aplicaciones o servicios que no están estandarizados, pero se registran para su uso por software y servicios específicos.

3. **Puertos dinámicos o privados (49152-65535)**: Son puertos temporales asignados de manera automática a las aplicaciones para la comunicación entre procesos.

### **¿Cómo Funcionan los Puertos?**
Cuando un dispositivo (por ejemplo, una computadora o servidor) recibe datos desde la red, utiliza una dirección IP junto con un número de puerto para determinar qué servicio o aplicación debe procesar esos datos. Este modelo se basa en el protocolo de comunicación **TCP/IP**.

Por ejemplo:
- Una solicitud de un navegador web (HTTP) utiliza el **puerto 80**.
- Si un servidor web recibe datos en el puerto 80, sabe que debe procesarlos como solicitudes HTTP.

### **Ejemplos de Puertos Comunes**

| **Puerto** | **Protocolo**  | **Descripción**                                   |
|------------|----------------|---------------------------------------------------|
| 22         | SSH            | Protocolo seguro para administración remota       |
| 23         | Telnet         | Comunicación en texto sin cifrado (obsoleto)      |
| 25         | SMTP           | Envío de correos electrónicos                     |
| 53         | DNS            | Resolución de nombres de dominio                 |
| 80         | HTTP           | Protocolo para navegar en la web                  |
| 443        | HTTPS          | HTTP seguro (cifrado)                             |
| 3306       | MySQL          | Acceso a bases de datos MySQL                     |
| 3389       | RDP            | Protocolo de escritorio remoto (Windows)          |
| 8080       | HTTP alternativo| Utilizado por servidores web como puerto secundario |

### **Vulnerabilidades Relacionadas con los Puertos**
Cada puerto que esté abierto en un sistema o dispositivo representa un posible **punto de entrada** para un atacante. Si un puerto no está bien protegido, los atacantes pueden aprovecharse de vulnerabilidades en los servicios que se ejecutan en ese puerto.

Las vulnerabilidades de los puertos pueden manifestarse de diversas maneras:
1. **Puertos Abiertos no Necesarios**:
   Si un puerto está abierto sin necesidad (por ejemplo, un servicio no utilizado), se convierte en una puerta abierta para los atacantes. Un buen ejemplo es tener puertos como **Telnet (23)** o **FTP (21)** habilitados, que no son seguros y pueden ser explotados.

2. **Explotación de Vulnerabilidades en Servicios**:
   Si un servicio en un puerto tiene una vulnerabilidad conocida (como una **inyección de SQL** en un puerto 3306 para MySQL o una **explotación de desbordamiento de búfer** en un servicio SSH), un atacante puede aprovecharla para acceder al sistema.

3. **Falta de Cifrado**:
   Algunos servicios, como **Telnet** o **FTP**, no cifran los datos. Esto permite que un atacante escuche el tráfico en la red y obtenga información sensible como contraseñas.

4. **Denegación de Servicio (DoS)**:
   Un puerto específico puede ser objetivo de un ataque de **denegación de servicio** (DoS), en el cual se inundan las conexiones hacia el puerto con tráfico falso, impidiendo que los servicios legítimos accedan al puerto.

### **Cómo Identificar Vulnerabilidades en los Puertos**

Para identificar vulnerabilidades en los puertos de una máquina o red, se pueden realizar varios pasos:

#### **1. Escaneo de Puertos con Herramientas**
El primer paso es **escuchar** los puertos abiertos en un sistema o red. Esto se puede hacer con herramientas como **Nmap**, que permite hacer un escaneo de puertos y descubrir qué servicios están corriendo en esos puertos. Por ejemplo:

```bash
nmap -p 1-65535 <ip_del_objetivo>
```

Este comando escanea todos los puertos de un objetivo (especificado por su dirección IP) para ver qué puertos están abiertos.

#### **2. Uso de Escáneres de Vulnerabilidades**
Una vez identificados los puertos abiertos, podemos usar herramientas de escaneo de vulnerabilidades como **Nessus** o **OpenVAS** para verificar si esos puertos están siendo utilizados por servicios con vulnerabilidades conocidas.

#### **3. Pruebas de Exploitación**
Una vez detectados servicios vulnerables en ciertos puertos, se pueden usar **exploit frameworks** como **Metasploit** para intentar explotar esas vulnerabilidades y obtener acceso no autorizado al sistema. Aunque esto debe hacerse solo con fines de prueba de penetración en entornos controlados, es útil para evaluar la seguridad de los puertos.

#### **4. Análisis de Logs y Tráfico**
El monitoreo de tráfico de red a través de herramientas como **Wireshark** o **tcpdump** puede ayudar a detectar actividades sospechosas que involucran puertos específicos. Los registros (logs) de los firewalls también pueden ser útiles para identificar intentos de acceso no autorizado a puertos específicos.

### **Ejemplo Práctico: Identificación de Vulnerabilidad en un Puerto con Nmap**

1. **Escaneo de Puertos**:
   Supongamos que queremos escanear los puertos de una dirección IP, como `192.168.1.10`. Usamos Nmap para escanear los puertos más comunes:

   ```bash
   nmap 192.168.1.10
   ```

   Nmap devolverá una lista de puertos abiertos y los servicios asociados.

2. **Análisis de Vulnerabilidades**:
   Una vez que tengamos la lista de puertos abiertos, por ejemplo, el puerto 80 (HTTP) o el puerto 22 (SSH), debemos analizar si los servicios tienen vulnerabilidades conocidas. Herramientas como **Nessus** o **Nikto** pueden ser utilizadas para escanear estos servicios y buscar vulnerabilidades.

3. **Simulación de un Ataque**:
   Si identificamos que un puerto está usando un servicio vulnerable, como un servidor web desactualizado en el puerto 80, podemos intentar explotar esa vulnerabilidad usando herramientas como **Metasploit**.

### **Conclusión**

La seguridad de los puertos es un aspecto crítico en la ciberseguridad, ya que cualquier puerto abierto puede representar un riesgo potencial. El análisis regular de puertos, el uso de firewalls, y el escaneo de vulnerabilidades son pasos esenciales para proteger las redes y sistemas. A medida que avanzas en el campo de la ciberseguridad, estas prácticas te ayudarán a asegurar que solo los servicios necesarios estén habilitados y protegidos, mientras se mitigan los riesgos de vulnerabilidades de puertos.

### **Paso a Paso: Identificar Puertos Abiertos o en Uso en tu Computador con Windows**

En Windows, puedes identificar qué puertos están abiertos o en uso mediante herramientas integradas, como **netstat** y **PowerShell**, así como mediante herramientas de terceros como **Nmap**. A continuación, te explico cómo hacerlo en detalle.

---

### **Método 1: Usando el Comando `netstat`**

**netstat** es una herramienta de línea de comandos que te permite ver las conexiones de red activas en tu computadora, así como los puertos que están en uso.

#### **Paso 1: Abrir la Consola de Comandos (CMD)**

1. Haz clic en el botón de **Inicio**.
2. Escribe **cmd** o **Símbolo del sistema** en la barra de búsqueda.
3. Haz clic con el botón derecho y selecciona **Ejecutar como administrador**.

#### **Paso 2: Ejecutar el Comando `netstat`**

Para ver las conexiones activas y los puertos abiertos, ejecuta el siguiente comando:

```bash
netstat -an | findstr LISTEN
```

**Explicación**:
- **`-a`**: Muestra todas las conexiones y puertos en uso.
- **`-n`**: Muestra las direcciones IP y puertos en formato numérico (sin resolver los nombres).
- **`findstr LISTEN`**: Filtra la salida para mostrar solo los puertos que están en estado **LISTEN**, es decir, los puertos en espera de conexiones entrantes.

#### **Paso 3: Interpretar los Resultados**

El comando te mostrará una lista de puertos abiertos y en escucha. La salida tendrá un formato como el siguiente:

```
TCP    0.0.0.0:80              0.0.0.0:0              LISTENING
TCP    [::]:443                [::]:0                  LISTENING
```

En este ejemplo:
- El puerto **80** está abierto y en espera de conexiones en todas las interfaces de red (0.0.0.0).
- El puerto **443** está abierto para conexiones seguras (HTTPS).

#### **Paso 4: Verificar Qué Servicio Está Usando un Puerto Específico**

Si deseas saber qué aplicación está usando un puerto en específico, puedes usar el siguiente comando para obtener el **PID (ID del proceso)** de las conexiones activas:

```bash
netstat -ano | findstr LISTEN
```

Esto añadirá una columna al final con el **PID** del proceso que está utilizando el puerto.

Luego, para identificar qué aplicación corresponde a ese PID, abre el **Administrador de tareas**:
1. Haz clic derecho en la barra de tareas y selecciona **Administrador de tareas**.
2. Ve a la pestaña **Detalles**.
3. Busca el **PID** que has encontrado en el comando `netstat` para identificar el proceso.

---

### **Método 2: Usando PowerShell**

**PowerShell** también ofrece una forma sencilla de ver los puertos abiertos en Windows.

#### **Paso 1: Abrir PowerShell**

1. Haz clic en el botón de **Inicio**.
2. Escribe **PowerShell** en la barra de búsqueda.
3. Haz clic derecho y selecciona **Ejecutar como administrador**.

#### **Paso 2: Ejecutar el Comando en PowerShell**

Ejecuta el siguiente comando para obtener una lista de los puertos abiertos y las conexiones:

```powershell
Get-NetTCPConnection | Where-Object {$_.State -eq "Listen"}
```

Esto mostrará una lista de las conexiones TCP que están en estado **LISTEN** (en espera de conexión). El comando también muestra la dirección local y el puerto asociado.

#### **Paso 3: Verificar Detalles del Proceso**

Para ver más detalles sobre el proceso que está utilizando un puerto, puedes usar el siguiente comando, que combina la información de **`Get-NetTCPConnection`** con el **`Get-Process`**:

```powershell
Get-NetTCPConnection -State Listen | Select-Object LocalAddress, LocalPort, OwningProcess | ForEach-Object { $_ | Add-Member -MemberType NoteProperty -Name ProcessName -Value ((Get-Process -Id $_.OwningProcess).Name) -Force; $_ } | Format-Table -AutoSize
```

Este comando no solo muestra el puerto y el proceso asociado, sino también el nombre del proceso que está utilizando ese puerto.

---

### **Método 3: Usando Nmap (Herramienta de Terceros)**

**Nmap** es una herramienta muy poderosa para escanear puertos en tu máquina o en una red. Puedes usarla para obtener una lista detallada de puertos abiertos y sus servicios.

#### **Paso 1: Descargar Nmap**

1. Dirígete al sitio web oficial de **Nmap**: [https://nmap.org/download.html](https://nmap.org/download.html).
2. Descarga e instala **Nmap** para Windows.

#### **Paso 2: Ejecutar Nmap**

1. Una vez instalado, abre una ventana de **cmd** o **PowerShell**.
2. Ejecuta el siguiente comando para escanear los puertos de tu máquina local (suponiendo que tu IP local es `127.0.0.1`):

```bash
nmap 127.0.0.1
```

Nmap escaneará todos los puertos de la máquina local y te mostrará los puertos abiertos, junto con los servicios asociados.

#### **Paso 3: Ejemplo de Salida de Nmap**

El comando puede mostrar una salida como esta:

```
Starting Nmap 7.92 ( https://nmap.org ) at 2025-01-13 12:00 UTC
Nmap scan report for 127.0.0.1
Host is up (0.00045s latency).
Not shown: 999 closed ports
PORT    STATE SERVICE
22/tcp  open  ssh
80/tcp  open  http
443/tcp open  https
8080/tcp open  http-proxy
```

Esta salida indica que los puertos **22 (SSH)**, **80 (HTTP)**, **443 (HTTPS)** y **8080 (HTTP Proxy)** están abiertos en la máquina local.

### **Método 4: Usando el Firewall de Windows para Ver Puertos Abiertos**

El **Firewall de Windows** también puede mostrarte las conexiones entrantes y los puertos que están permitidos.

#### **Paso 1: Abrir el Firewall de Windows**

1. Ve al **Panel de control**.
2. Haz clic en **Firewall de Windows Defender**.
3. En el panel izquierdo, haz clic en **Configuración avanzada**.

#### **Paso 2: Ver Reglas de Entrada**

En la ventana de configuración avanzada, puedes ver las **Reglas de Entrada** que determinan qué puertos están permitidos en tu sistema. Aquí puedes revisar puertos específicos que están habilitados para acceder a tu máquina.

### **Resumen**

- **netstat** y **PowerShell** son herramientas integradas en Windows que te permiten ver los puertos abiertos en tu máquina.
- **Nmap** es una excelente herramienta de terceros que ofrece un análisis más detallado y flexible.
- Puedes combinar estas herramientas con un análisis de procesos para identificar qué aplicaciones están utilizando cada puerto.
- Además, puedes revisar las reglas del **Firewall de Windows** para asegurarte de que solo los puertos necesarios estén habilitados.
