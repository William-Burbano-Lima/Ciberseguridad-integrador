Ejercicio de **phishing** en entornos controlados utilizando **Social Engineering Toolkit (SET)**
___

El **Social Engineering Toolkit (SET)** es una de las herramientas más poderosas utilizadas para realizar ataques de ingeniería social, que simulan ataques reales de phishing y otras formas de manipulación de usuarios para obtener información confidencial. A continuación, te proporcionaré un paso a paso detallado sobre cómo usar SET en **Kali Linux** para realizar un ataque de **phishing**.

Recuerda, este tipo de simulaciones debe hacerse únicamente con fines educativos, en un entorno controlado y con permiso explícito de las partes involucradas.

---

### **Paso 1: Instalación y Preparación de Kali Linux**

1. **Actualizar Kali Linux**:
   Antes de usar SET, asegúrate de que tu sistema esté completamente actualizado. Abre una terminal y ejecuta los siguientes comandos:
   ```bash
   sudo apt update && sudo apt upgrade -y
   ```

2. **Instalar Social Engineering Toolkit (SET)**:
   Si no tienes SET instalado, puedes instalarlo ejecutando:
   ```bash
   sudo apt install set
   ```

3. **Verificar la instalación de SET**:
   Después de la instalación, asegúrate de que SET está correctamente instalado ejecutando:
   ```bash
   setoolkit
   ```

   Esto iniciará el **Social Engineering Toolkit**.

---

### **Paso 2: Iniciar SET**

1. **Iniciar SET**:
   Para comenzar a trabajar con SET, abre una terminal y ejecuta:
   ```bash
   sudo setoolkit
   ```

2. **Menú Principal de SET**:
   Al ejecutarlo, aparecerá el menú principal de SET. Aquí, tendrás varias opciones disponibles para elegir, como ataques de phishing, ataques por USB, ataques de ingeniería social, etc.

---

### **Paso 3: Seleccionar el Tipo de Ataque**

1. **Seleccionar el tipo de ataque**:
   Una vez dentro de SET, verás varias opciones. Para un ataque de phishing, selecciona la opción **"Social-Engineering Attacks"**. Generalmente, esta es la opción **1**.

   ```text
   1) Social-Engineering Attacks
   ```

2. **Seleccionar "Website Attack Vectors"**:
   Después de elegir "Social-Engineering Attacks", SET te ofrecerá diferentes vectores de ataque. Elige la opción **"Website Attack Vectors"**. Normalmente, esta es la opción **2**.

   ```text
   2) Website Attack Vectors
   ```

3. **Seleccionar "Credential Harvester Attack Method"**:
   Ahora, elige la opción **"Credential Harvester Attack Method"**. Este método está diseñado para capturar credenciales enviadas por las víctimas a través de un sitio web falso que se les presentará.

   ```text
   3) Credential Harvester Attack Method
   ```

---

### **Paso 4: Configuración del Ataque de Phishing**

1. **Seleccionar "Site Cloner"**:
   La siguiente opción que verás es para elegir el método que se usará para clonar el sitio web. Aquí, selecciona la opción **"Site Cloner"** (esto te permite clonar un sitio web legítimo, como Gmail o Facebook, y mostrarlo a la víctima).

   ```text
   1) Site Cloner
   ```

2. **Elegir el sitio web a clonar**:
   Ahora, SET te pedirá que elijas el sitio web que deseas clonar. Por ejemplo, si deseas clonar Gmail, escribe **gmail.com**. Asegúrate de que el sitio esté en la lista de sitios preconfigurados en SET.

   ```text
   Enter the URL to clone (example: gmail.com): gmail.com
   ```

3. **Configurar el servidor web**:
   A continuación, SET te pedirá que configures el servidor web para que se encargue de servir la página falsa. El servidor web se configurará automáticamente para que escuche en el puerto **80** (HTTP) o **443** (HTTPS).

   Asegúrate de que tu máquina tenga una dirección IP estática configurada en **Kali Linux**, como expliqué en la sección anterior. Si no tienes una IP estática, puedes configurar un dominio falso para redirigir a la víctima a tu máquina local utilizando la IP local de tu Kali.

4. **Generar y enviar el enlace**:
   Una vez que SET haya configurado todo, generará un enlace a la página de phishing, que será algo como:
   ```
   http://<tu-ip-local>/login
   ```

   **Ejemplo**: Si tu máquina Kali tiene la IP **192.168.1.100**, el enlace será:
   ```
   http://192.168.1.100/login
   ```

---

### **Paso 5: Capturar las Credenciales**

1. **Enviar el enlace a la víctima**:
   Ahora que tienes el enlace generado, puedes enviarlo a la víctima a través de diferentes canales (correo electrónico, mensajes de texto, etc.), haciéndolo parecer un enlace legítimo, como una solicitud de actualización de contraseña.

2. **La víctima ingresa sus credenciales**:
   Cuando la víctima haga clic en el enlace y se le presente la página clonada (pareciendo ser Gmail o el sitio legítimo), ingresará su nombre de usuario y contraseña en el formulario.

3. **Captura de credenciales**:
   Una vez que la víctima haya ingresado sus credenciales, estas serán enviadas al servidor de SET y aparecerán en la terminal de Kali Linux. Puedes verlas en tiempo real mientras la víctima ingresa sus datos.

   Las credenciales serán capturadas por SET y mostradas de manera similar a esto:
   ```text
   Credenciales Capturadas:
   -------------------------
   Usuario: victim@example.com
   Contraseña: password123
   ```

---

### **Paso 6: Finalizar y Analizar**

1. **Detener el servidor**:
   Cuando hayas terminado con la simulación, puedes detener el servidor de SET presionando **Ctrl+C** en la terminal donde lo ejecutaste.

2. **Revisar las credenciales**:
   Las credenciales de las víctimas se guardan en el directorio **/tmp/.set/**, y también se muestran en tiempo real en la terminal.

---

### **Consejos Adicionales para Seguridad y Prevención:**

Aunque este es un ataque simulado, es importante comprender cómo defenderse de ataques de phishing en la vida real. Algunos consejos para protegerse incluyen:

- **Revisar siempre la URL**: Verifica que la URL sea legítima y comience con "https://" antes de ingresar cualquier dato sensible.
- **Habilitar la autenticación de dos factores (2FA)**: Esto ayuda a proteger las cuentas incluso si un atacante obtiene las credenciales.
- **Nunca hacer clic en enlaces desconocidos**: Si no esperas un correo electrónico o mensaje, no hagas clic en enlaces o descarges archivos de fuentes desconocidas.
- **Educación continua**: La mejor defensa contra ataques de ingeniería social es estar informado y educado sobre las amenazas actuales.
