## 🔍 **PRÁCTICA 4: Análisis de Tráfico con Wireshark**  
📌 **Objetivo:** Identificar patrones sospechosos en el tráfico de red y detectar posibles ataques.  

### **Instrucciones:**  
1. **Escenario:**  
   - Un usuario reporta que su sistema se ha vuelto lento después de recibir un correo sospechoso.  
   - Se sospecha de un posible ataque de **phishing o malware**.  
   - Se usará **Wireshark** para analizar paquetes de red.  

2. **Paso 1 - Captura de Tráfico:**  
   - **Abrir Wireshark** y seleccionar la interfaz de red activa.  
   - Iniciar la captura y navegar por una página web conocida (ej. un sitio seguro).  
   - Detener la captura tras unos segundos.  

3. **Paso 2 - Análisis de Tráfico Sospechoso:**  
   - Filtrar tráfico por **protocolo DNS**:  
     ```
     dns
     ```
   - Buscar solicitudes a **dominios desconocidos**.  
   - Revisar tráfico **HTTP o HTTPS** en busca de **URLs sospechosas** con:  
     ```
     http contains "password"
     ```
   - Detectar intentos de escaneo de puertos con:  
     ```
     tcp.flags.syn == 1 && tcp.flags.ack == 0
     ```
  
4. **Paso 3 - Informe de Auditoría:**  
   - Llenar una **Matriz de Análisis Wireshark** con hallazgos:  

| **Tipo de Tráfico** | **IP de Origen** | **IP de Destino** | **Acción Correctiva** |
|------------------|------------------|------------------|------------------|
| DNS sospechoso | 192.168.1.10 | 185.45.13.22 | Bloquear dominio |
| HTTP sospechoso | 192.168.1.12 | 203.0.113.5 | Inspección de correo |
| Escaneo de puertos | 192.168.1.30 | 10.0.0.1 | Bloquear IP |

🔍 **Resultado esperado:**  
- Detectar patrones de tráfico malicioso y generar reportes de auditoría.

## 📊 **PRÁCTICA 5: Análisis de Logs con Splunk**  
📌 **Objetivo:** Usar **Splunk** para analizar logs de seguridad y detectar eventos críticos.  

### **Instrucciones:**  
1. **Escenario:**  
   - Una empresa usa **Splunk** para gestionar eventos de seguridad.  
   - Se sospecha de **accesos no autorizados** en los servidores.  

2. **Paso 1 - Carga de Logs en Splunk:**  
   - Ingresar a Splunk y cargar logs de ejemplo.  
   - Usar consultas para detectar accesos sospechosos:  
     ```
     index=security sourcetype=access_logs "failed login"
     ```
   - Identificar múltiples intentos de acceso fallidos desde la misma IP:  
     ```
     index=security sourcetype=access_logs | stats count by src_ip
     ```

3. **Paso 2 - Creación de Dashboard de Seguridad:**  
   - Graficar **accesos fallidos** por país.  
   - Configurar **alertas** para más de 5 intentos fallidos en 1 minuto.  

4. **Paso 3 - Matriz de Acciones Correctivas:**  

| **Evento Detectado** | **IP Origen** | **Hora** | **Medida Correctiva** |
|------------------|------------------|------------------|------------------|
| Intentos de acceso fallidos | 103.22.12.5 | 02:15 AM | Bloqueo IP |
| Acceso exitoso fuera de horario | 192.168.2.15 | 03:00 AM | Notificación SOC |
| Intento de escalada de privilegios | 10.0.0.3 | 04:30 PM | Auditoría interna |

🔍 **Resultado esperado:**  
- Monitoreo en tiempo real de accesos sospechosos con Splunk.

## 🛠 **PRÁCTICA 6: Ethical Hacking con Kali Linux**  
📌 **Objetivo:** Simular un ataque de escaneo de puertos con **Nmap** y generar un plan de remediación.  

### **Instrucciones:**  
1. **Escenario:**  
   - Un pentester usa **Kali Linux** para auditar un servidor en busca de vulnerabilidades.  
   - Se debe **simular un escaneo de red y analizar los resultados**.  

2. **Paso 1 - Escaneo de Puertos con Nmap:**  
   - En Kali Linux, ejecutar:  
     ```
     nmap -sV -p- 192.168.1.1
     ```
   - Identificar servicios abiertos y vulnerabilidades.  

3. **Paso 2 - Análisis de Resultados:**  
   - Si se detectan puertos abiertos sospechosos, anotar en la matriz:  

| **Puerto** | **Servicio Detectado** | **Riesgo** | **Recomendación** |
|-----------|----------------------|-----------|------------------|
| 22 | OpenSSH 7.2 | Medio | Actualizar versión |
| 3389 | RDP | Alto | Restringir acceso remoto |
| 445 | SMB | Crítico | Deshabilitar si no es necesario |

4. **Paso 3 - Plan de Remediación:**  
   - Implementar medidas como **firewalls, restricciones de acceso y actualizaciones**.  

🔍 **Resultado esperado:**  
- Detectar servicios vulnerables y aplicar medidas de mitigación.
