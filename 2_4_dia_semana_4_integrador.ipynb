## 沐 **PRﾃ，TICA 4: Anﾃ｡lisis de Trﾃ｡fico con Wireshark**  
沒 **Objetivo:** Identificar patrones sospechosos en el trﾃ｡fico de red y detectar posibles ataques.  

### **Instrucciones:**  
1. **Escenario:**  
   - Un usuario reporta que su sistema se ha vuelto lento despuﾃｩs de recibir un correo sospechoso.  
   - Se sospecha de un posible ataque de **phishing o malware**.  
   - Se usarﾃ｡ **Wireshark** para analizar paquetes de red.  

2. **Paso 1 - Captura de Trﾃ｡fico:**  
   - **Abrir Wireshark** y seleccionar la interfaz de red activa.  
   - Iniciar la captura y navegar por una pﾃ｡gina web conocida (ej. un sitio seguro).  
   - Detener la captura tras unos segundos.  

3. **Paso 2 - Anﾃ｡lisis de Trﾃ｡fico Sospechoso:**  
   - Filtrar trﾃ｡fico por **protocolo DNS**:  
     ```
     dns
     ```
   - Buscar solicitudes a **dominios desconocidos**.  
   - Revisar trﾃ｡fico **HTTP o HTTPS** en busca de **URLs sospechosas** con:  
     ```
     http contains "password"
     ```
   - Detectar intentos de escaneo de puertos con:  
     ```
     tcp.flags.syn == 1 && tcp.flags.ack == 0
     ```
  
4. **Paso 3 - Informe de Auditorﾃｭa:**  
   - Llenar una **Matriz de Anﾃ｡lisis Wireshark** con hallazgos:  

| **Tipo de Trﾃ｡fico** | **IP de Origen** | **IP de Destino** | **Acciﾃｳn Correctiva** |
|------------------|------------------|------------------|------------------|
| DNS sospechoso | 192.168.1.10 | 185.45.13.22 | Bloquear dominio |
| HTTP sospechoso | 192.168.1.12 | 203.0.113.5 | Inspecciﾃｳn de correo |
| Escaneo de puertos | 192.168.1.30 | 10.0.0.1 | Bloquear IP |

沐 **Resultado esperado:**  
- Detectar patrones de trﾃ｡fico malicioso y generar reportes de auditorﾃｭa.

## 沒 **PRﾃ，TICA 5: Anﾃ｡lisis de Logs con Splunk**  
沒 **Objetivo:** Usar **Splunk** para analizar logs de seguridad y detectar eventos crﾃｭticos.  

### **Instrucciones:**  
1. **Escenario:**  
   - Una empresa usa **Splunk** para gestionar eventos de seguridad.  
   - Se sospecha de **accesos no autorizados** en los servidores.  

2. **Paso 1 - Carga de Logs en Splunk:**  
   - Ingresar a Splunk y cargar logs de ejemplo.  
   - Usar consultas para detectar accesos sospechosos:  
     ```
     index=security sourcetype=access_logs "failed login"
     ```
   - Identificar mﾃｺltiples intentos de acceso fallidos desde la misma IP:  
     ```
     index=security sourcetype=access_logs | stats count by src_ip
     ```

3. **Paso 2 - Creaciﾃｳn de Dashboard de Seguridad:**  
   - Graficar **accesos fallidos** por paﾃｭs.  
   - Configurar **alertas** para mﾃ｡s de 5 intentos fallidos en 1 minuto.  

4. **Paso 3 - Matriz de Acciones Correctivas:**

## 泝 **PRﾃ，TICA 6: Ethical Hacking con Kali Linux**  
沒 **Objetivo:** Simular un ataque de escaneo de puertos con **Nmap** y generar un plan de remediaciﾃｳn.  

### **Instrucciones:**

| 3389 | RDP | Alto | Restringir acceso remoto |
| 445 | SMB | Crﾃｭtico | Deshabilitar si no es necesario |

4. **Paso 3 - Plan de Remediaciﾃｳn:**  
   - Implementar medidas como **firewalls, restricciones de acceso y actualizaciones**.
