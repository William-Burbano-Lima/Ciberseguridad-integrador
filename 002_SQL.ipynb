¿Qué son las consultas SQL?

SQL, que significa Structured Query Language, es un lenguaje que se utiliza para comunicarse con bases de datos y manipularlas. Es el lenguaje estándar para los sistemas de gestión de bases de datos relacionales y se utiliza para realizar tareas como actualizar datos en una base de datos o recuperar datos de una base de datos.

Las consultas SQL son los comandos que se utilizan para comunicarse con una base de datos. Permiten realizar tareas como buscar, actualizar o recuperar datos almacenados en una base de datos. Por ejemplo, una aplicación web de comercio electrónico podría utilizar una consulta SQL para recuperar información sobre un producto específico de la base de datos de productos.

Sin embargo, si una aplicación no está diseñada de forma segura, puede permitir que un usuario ingrese consultas SQL directamente. En ese caso, un usuario malintencionado puede ingresar una consulta SQL especialmente diseñada que puede provocar acceso no autorizado, fuga de datos o incluso eliminación de datos. Esto se conoce como ataque de inyección SQL.
Impacto de un ataque de inyección SQL exitoso

Las consecuencias de un ataque de inyección SQL exitoso pueden incluir:

    Credenciales robadas : los atacantes pueden obtener credenciales a través de SQLi y luego hacerse pasar por usuarios y usar sus privilegios.
    Acceso no autorizado a bases de datos : los atacantes pueden obtener acceso a datos confidenciales en servidores de bases de datos.
    Alteración de datos : los atacantes pueden alterar o agregar nuevos datos a la base de datos a la que se accede.
    Eliminación de datos : los atacantes pueden eliminar registros de bases de datos o eliminar tablas enteras.

Movimiento lateral : los atacantes pueden acceder a servidores de bases de datos con privilegios de sistema operativo y usar estos permisos para acceder a otros sistemas sensibles.
¿Cómo funciona un ataque de inyección SQL?

Un ataque de inyección SQL implica insertar o “inyectar” una consulta SQL a través de los datos de entrada del cliente a la aplicación. Un ataque exitoso permite a un atacante manipular las consultas SQL que una aplicación realiza a su base de datos. Por lo general, implica los siguientes pasos:

**Identificación de entradas vulnerables:** los atacantes primero identifican las entradas dentro de la aplicación web que son vulnerables a la inyección SQL. Estas entradas pueden ser campos de texto en un formulario, parámetros de URL o cualquier otro mecanismo de entrada.

**Elaboración de la consulta SQL maliciosa:** una vez que se identifica una entrada vulnerable, los atacantes elaboran una sentencia SQL destinada a insertarse en la consulta ejecutada por la aplicación. Esta sentencia está diseñada para modificar la consulta SQL original y realizar acciones no previstas por los desarrolladores de la aplicación.

**Evitar las medidas de seguridad de la aplicación:** los atacantes a menudo tienen que eludir las medidas de seguridad, como la validación de entradas o el escape de caracteres especiales. Lo logran mediante técnicas como la concatenación de cadenas o el uso de la sintaxis SQL para comentar partes de la consulta original.

**Ejecución de la consulta maliciosa:** cuando la aplicación ejecuta la consulta SQL, incluye la entrada maliciosa del atacante. Esta consulta modificada puede realizar acciones como la visualización no autorizada de datos, la eliminación de datos o incluso la modificación del esquema de la base de datos.

**Extracción o manipulación de datos:** dependiendo del ataque, el resultado podría ser la extracción de información confidencial (como credenciales de usuario), la alteración de datos existentes, la adición de datos nuevos o incluso la eliminación de partes importantes de la base de datos.

**Explotación de vulnerabilidades del servidor de bases de datos:**
las inyecciones SQL avanzadas pueden explotar vulnerabilidades en el servidor de bases de datos, extendiendo el ataque más allá de la base de datos hasta el nivel del servidor. Esto puede incluir la ejecución de comandos en el sistema operativo o el acceso a otras partes del sistema de archivos del servidor.

Este proceso aprovecha la ejecución dinámica de SQL en aplicaciones en las que las entradas del usuario se incluyen directamente en las sentencias SQL sin la validación o el escape adecuados. Explota la forma en que se construyen las consultas SQL, a menudo de una manera que los desarrolladores no anticiparon.

## Tipos de ataques de inyección SQL

Existen varios tipos de inyección SQL:

**Inyección SQL basada en unión :** la inyección SQL basada en unión representa el tipo más popular de inyección SQL y utiliza la instrucción UNION. La instrucción UNION representa la combinación de dos instrucciones select para recuperar datos de la base de datos.

**Inyección SQL basada en errores :** este método solo se puede ejecutar contra servidores MS-SQL. En este ataque, el usuario malintencionado hace que una aplicación muestre un error. Por lo general, se le hace una pregunta a la base de datos y esta devuelve un mensaje de error que también contiene los datos solicitados.

**Inyección SQL ciega :** en este ataque, no se reciben mensajes de error de la base de datos; extraemos los datos enviando consultas a la base de datos. Las inyecciones SQL ciegas se pueden dividir en inyección SQL basada en booleanos e inyección SQL basada en tiempo.

Los ataques SQLi también se pueden clasificar según el método que utilizan para inyectar datos:

**Inyección SQL basada en la entrada del usuario :** las aplicaciones web aceptan entradas a través de formularios, que pasan la entrada del usuario a la base de datos para su procesamiento. Si la aplicación web acepta estas entradas sin desinfectarlas, un atacante puede inyectar sentencias SQL maliciosas.

**Inyección SQL basada en cookies :** otro enfoque de la inyección SQL consiste en modificar las cookies para “envenenar” las consultas de la base de datos. Las aplicaciones web suelen cargar cookies y utilizar sus datos como parte de las operaciones de la base de datos. Un usuario malintencionado o un malware instalado en el dispositivo de un usuario podrían modificar las cookies para inyectar SQL de forma inesperada.

**Inyección SQL basada en encabezados HTTP :** las variables del servidor, como los encabezados HTTP, también se pueden utilizar para la inyección SQL. Si una aplicación web acepta entradas de encabezados HTTP, los encabezados falsos que contienen SQL arbitrario pueden inyectar código en la base de datos.

**Inyección SQL de segundo orden :** estos son posiblemente los ataques de inyección SQL más complejos, porque pueden permanecer inactivos durante un largo período de tiempo. Un ataque de inyección SQL de segundo orden envía datos envenenados, que podrían considerarse benignos en un contexto, pero son maliciosos en otro. Incluso si los desarrolladores desinfectan todas las entradas de la aplicación, aún podrían ser vulnerables a este tipo de ataque.

## Ejemplos de código de inyección SQL

Veamos dos ejemplos comunes de ataques de inyección SQL.

**Ejemplo 1:** Uso de SQLi para autenticarse como administrador

Este ejemplo muestra cómo un atacante puede usar la inyección SQL para eludir la autenticación de una aplicación y obtener privilegios de administrador.

Considere un sistema de autenticación simple que utiliza una tabla de base de datos con nombres de usuario y contraseñas. La solicitud POST de un usuario proporcionará las variables usery pass, y estas se insertan en una declaración SQL:

**sql = "SELECT id FROM users WHERE username='" + user + "' AND password='" + pass + "'"**

El problema aquí es que la sentencia SQL utiliza la concatenación para combinar datos. El atacante puede proporcionar una cadena como esta en lugar de la passvariable:

password' OR 5=5

La consulta SQL resultante se ejecutará en la base de datos:

SELECT id FROM users WHERE username='user' AND password='pass' OR 5=5'

Debido a que 5=5es una condición que siempre se evalúa como verdadera, toda la WHEREdeclaración será verdadera, independientemente del nombre de usuario o la contraseña proporcionados.

La WHEREdeclaración devolverá el primer ID de la tabla de usuarios, que normalmente es el del administrador. Esto significa que el atacante puede acceder a la aplicación sin autenticación y también tiene privilegios de administrador.

Una forma más avanzada de este ataque es cuando el atacante agrega un símbolo de comentario de código al final de la declaración SQL, lo que le permite manipular aún más la consulta SQL. Lo siguiente funcionará en la mayoría de las bases de datos, incluidas MySQL, PostgreSQL y Oracle:

**' OR '5'='5' /***


**Ejemplo 2:** Uso de SQLi para acceder a datos confidenciales

En este ejemplo, el siguiente código obtiene el nombre de usuario actual y busca elementos que coincidan con un determinado nombre de elemento, donde el propietario es el usuario actual.

...
string userName = ctx.getAuthenticatedUserName();
string query = "SELECT * FROM items WHERE owner = "'"
                + userName + "' AND itemname = '"
                + ItemName.Text + "'";
...

Este código tiene la misma debilidad que el ejemplo anterior: el uso de la concatenación. Después de combinar el nombre de usuario y el nombre del elemento, el código crea la siguiente consulta:

SELECT * FROM items
WHERE owner =
AND itemname = ;

Si el atacante proporciona la siguiente cadena para itemname:

Widget' OR 5=5

La declaración SQL se convierte en:

SELECT * FROM items
WHERE owner = 'John'
AND itemname = 'Widget' OR 5=5';

Que es lo mismo que: SELECT * FROM items;

Esto significa que la consulta devolverá los datos de toda la tabla, lo que le dará al atacante acceso no autorizado a datos confidenciales.

**Ejemplo 3:** Inyección de declaraciones maliciosas en un campo de formulario

Se trata de un ataque de inyección SQL simple basado en la entrada del usuario. El atacante utiliza un formulario que requiere el nombre y el apellido como datos de entrada. El atacante introduce:

    Nombre de pila:malicious'ex
    Apellido:Smith

La variable de nombre del atacante contiene una expresión maliciosa, a la que llamamos "ex". La instrucción SQL que procesa las entradas del formulario tiene el siguiente aspecto:

SELECT id, firstname, lastname FROM authors

Una vez que el atacante inyecta una expresión maliciosa en el nombre, la declaración se ve así:

SELECT id, firstname, lastname FROM authors WHERE firstname = 'malicious'ex' and lastname ='newman'

La base de datos identifica una sintaxis incorrecta debido al apóstrofe único e intenta ejecutar la declaración maliciosa.

**Opción 1 de defensa:** Declaraciones preparadas (con consultas parametrizadas)

Las sentencias preparadas son fáciles de aprender y usar, y eliminan el problema de la inyección SQL. Te obligan a definir el código SQL y a pasar cada parámetro a la consulta más tarde, lo que hace una clara distinción entre el código y los datos.

Si un atacante proporciona una cadena maliciosa como en los ejemplos anteriores, por ejemplo, si proporciona John' or 1=1un nombre de usuario, la declaración preparada la evaluará como una cadena literal. Buscará un usuario llamado John' or 1=1(y fallará, porque no existe dicho usuario) en lugar de evaluar esta declaración como código.

Las sentencias preparadas están disponibles en todos los lenguajes de programación. A continuación se muestra un ejemplo en Java. Para mayor seguridad, OWASP recomienda validar el parámetro de entrada por si acaso.

// Separate definition of input variable
String custname = request.getParameter("customerName");
// Separate definition of SQL statement
String query = "SELECT account_balance FROM user_data WHERE user_name = ? ";
// PreparedStatement command securely combines inputs and SQL syntax
PreparedStatement pstmt = connection.prepareStatement( query );
pstmt.setString( 1, custname);
ResultSet results = pstmt.executeQuery( );
Opción de defensa 2: procedimientos almacenados

Los procedimientos almacenados son similares a las sentencias preparadas, solo que el código SQL para el procedimiento almacenado se define y almacena en la base de datos, en lugar de en el código del usuario. En la mayoría de los casos, los procedimientos almacenados pueden ser tan seguros como las sentencias preparadas, por lo que puede decidir cuál se adapta mejor a sus procesos de desarrollo.

Hay dos casos en los que los procedimientos almacenados no son seguros:

El procedimiento almacenado incluye la generación de SQL dinámico ; esto no suele hacerse en los procedimientos almacenados, pero se puede hacer, por lo que debe evitarlo al crear procedimientos almacenados. De lo contrario, asegúrese de validar todas las entradas.

**Privilegios del propietario de la base de datos :** en algunas configuraciones de bases de datos, el administrador otorga permisos al propietario de la base de datos para permitir que se ejecuten los procedimientos almacenados. Esto significa que, si un atacante viola el servidor, tendrá todos los derechos sobre la base de datos. Evite esto creando un rol personalizado que permita a los procedimientos de almacenamiento solo el nivel de acceso que necesitan.
