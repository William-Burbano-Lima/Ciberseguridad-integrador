## **1. Pilares de la Ciberseguridad**  
Los tres pilares fundamentales garantizan la protección de la información en cualquier sistema:  

### **Confidencialidad**  
Se refiere a la protección de los datos para evitar que personas no autorizadas accedan a ellos.  
**Ejemplo:**  
- Un sistema bancario cifra las contraseñas de los usuarios para que nadie, ni siquiera los empleados del banco, puedan verlas.  

**Técnicas utilizadas:**  
- Cifrado de datos (AES, RSA).  
- Control de acceso basado en roles (RBAC).  
- Autenticación de dos factores (2FA).  

### **Integridad**  
Asegura que los datos no sean modificados sin autorización.  
**Ejemplo:**  
- Un correo electrónico firmado digitalmente garantiza que el mensaje no fue alterado durante su envío.  

**Técnicas utilizadas:**  
- Algoritmos de hash (SHA256, MD5).  
- Firmas digitales y certificados SSL.  
- Listas de control de acceso (ACLs).  

### **Disponibilidad**  
Garantiza que los datos y sistemas estén accesibles cuando sean necesarios.  
**Ejemplo:**  
- Un sitio web de comercio electrónico tiene servidores redundantes para seguir operando incluso si uno falla.  

**Técnicas utilizadas:**  
- Uso de balanceadores de carga.  
- Protección contra ataques DDoS.  
- Respaldos y redundancia de datos.

## **2. Vectores de Ataque**  
Son las formas en que un atacante puede comprometer un sistema.  

### **Ingeniería social**  
Manipulación psicológica para engañar a las personas y obtener información.  
**Ejemplo:**  
- Un atacante finge ser un técnico de soporte y convence a un empleado de la empresa para que le revele su contraseña.  

### **Phishing**  
Un tipo de ataque de ingeniería social donde se envían correos falsos para robar información.  
**Ejemplo:**  
- Un usuario recibe un correo que parece ser de su banco pidiéndole que ingrese su contraseña en un sitio web falso.  

### **Malware**  
Software malicioso diseñado para dañar o infiltrarse en un sistema.  
**Ejemplo:**  
- Un usuario descarga un archivo adjunto en un correo electrónico y su computadora se infecta con ransomware, bloqueando sus archivos.  

### **MITM (Man-in-the-Middle)**  
El atacante intercepta la comunicación entre dos partes sin que estas lo sepan.  
**Ejemplo:**  
- Un atacante se conecta a una red WiFi pública y captura los datos bancarios de los usuarios que inician sesión en su banca en línea.

## **3. Cifrado y Hashing**  
Técnicas para proteger datos mediante la transformación de la información.  

### **Cifrado Simétrico**  
Usa una sola clave para cifrar y descifrar datos.  
**Ejemplo:**  
- El protocolo WPA2 usa cifrado AES para proteger las redes WiFi.  

### **Cifrado Asimétrico**  
Usa un par de claves: una pública para cifrar y una privada para descifrar.  
**Ejemplo:**  
- Cuando inicias sesión en un sitio web con HTTPS, tu navegador usa cifrado asimétrico para proteger la conexión.  

### **Funciones Hash**  
Convierte datos en una huella digital única de longitud fija.  
**Ejemplo:**  
- Cuando descargas un software, su hash SHA256 se publica para que puedas verificar que el archivo no ha sido modificado.

## **4. Regulaciones y Derechos en Ciberseguridad**  

### **GDPR (Reglamento General de Protección de Datos)**  
Protege los datos personales de los ciudadanos de la UE.  
**Ejemplo:**  
- Un usuario puede solicitar que una empresa elimine todos sus datos personales de su base de datos.  

### **MSPI (Marco de Seguridad de la Información)**  
Define políticas y estándares de seguridad en organizaciones gubernamentales.  
**Ejemplo:**  
- Un organismo gubernamental establece reglas para el almacenamiento seguro de datos sensibles.

## **5. Seguridad en Redes y Protocolos**  

### **HTTP vs HTTPS**  
- **HTTP**: No cifra los datos, permitiendo que un atacante pueda interceptarlos.  
- **HTTPS**: Usa cifrado SSL/TLS para proteger la información.  

**Ejemplo:**  
- Un usuario introduce su tarjeta de crédito en un sitio con HTTP, y un atacante en la misma red puede robar la información.  

### **Firewalls**  
Dispositivos que filtran el tráfico de red según reglas predefinidas.  
**Ejemplo:**  
- Un firewall bloquea intentos de conexión sospechosos en un servidor web.  

### **VPN (Virtual Private Network)**  
Crea un túnel cifrado para la comunicación en internet.  
**Ejemplo:**  
- Un empleado que trabaja remotamente usa una VPN para conectarse de forma segura a la red de su empresa.

## **6. Errores en HTTP**  

- **Código 200**: Respuesta satisfactoria.  
- **Código 400**: Error del cliente (ejemplo: URL mal escrita).  
- **Código 500**: Error del servidor (ejemplo: fallo en la base de datos).  

**Ejemplo:**  
- Si intentas acceder a un sitio y ves un error "404 Not Found", significa que la página no existe.

## **7. Seguridad en Aplicaciones y Autenticación**  

### **OAuth2**  
Protocolo que permite a los usuarios autorizar aplicaciones sin compartir sus credenciales.  
**Ejemplo:**  
- Usar tu cuenta de Google para iniciar sesión en Spotify sin compartir tu contraseña.  

### **OpenID Connect**  
Extensión de OAuth2 para la autenticación de usuarios.  
**Ejemplo:**  
- Un usuario se autentica en una plataforma con su cuenta de Facebook sin escribir una contraseña.  

### **Hardening**  
Proceso de mejora de seguridad de un sistema.  
**Ejemplo:**  
- Un administrador de sistemas desactiva servicios innecesarios y cambia configuraciones predeterminadas para reducir vulnerabilidades.

## **8. Ataques Cibernéticos**  

### **DoS (Denegación de Servicio)**  
Satura un sistema con solicitudes hasta dejarlo inoperativo.  
**Ejemplo:**  
- Un sitio web recibe millones de peticiones falsas y deja de responder.  

### **DDoS (Denegación de Servicio Distribuida)**  
Lo mismo que DoS, pero usando múltiples dispositivos.  
**Ejemplo:**  
- Una botnet de miles de computadoras infectadas realiza un ataque DDoS contra un servidor bancario.  

### **Keylogger**  
Software que registra las teclas presionadas en un teclado.  
**Ejemplo:**  
- Un atacante instala un keylogger en una computadora pública y roba credenciales de acceso a cuentas bancarias.

## **9. Modelos y Organizaciones de Seguridad**  

### **Modelo OSI**  
Define 7 capas para la comunicación en redes (Física, Enlace de Datos, Red, Transporte, Sesión, Presentación y Aplicación).  
**Ejemplo:**  
- Un firewall opera en la capa 3 (Red) y la capa 4 (Transporte) para bloquear tráfico malicioso.  

### **OWASP (Open Web Application Security Project)**  
Organización que identifica vulnerabilidades en aplicaciones web.  
**Ejemplo:**  
- La lista OWASP Top 10 enumera las vulnerabilidades más comunes, como inyección SQL y Cross-Site Scripting (XSS).  

### **Hacker de Sombrero Blanco**  
Especialista en ciberseguridad que busca fallas en sistemas para protegerlos.  
**Ejemplo:**  
- Un hacker ético realiza pruebas de penetración en una empresa para identificar vulnerabilidades antes de que los atacantes lo hagan.
